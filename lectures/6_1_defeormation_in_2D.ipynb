## Poisson's number

Poisson's number (also Poisson's ratio) is a dimensionless material constant that describes the relationship between the transverse and longitudinal deformation of a material under mechanical loading. It indicates the ratio of the relative change in the width of the material to the relative change in length under tensile or compressive stress.

> Poisson's number $\nu$ is defined as the negative ratio of the transverse relative deformation $\varepsilon_{\perp}$ to the longitudinal relative deformation $\varepsilon_{\parallel}$:

$$
\nu = -\frac{\varepsilon_{\perp}}}{\varepsilon_{\parallel}}
$$

where:
- $\varepsilon_{\perp}$ is the relative deformation in the direction perpendicular to the loading direction,
- $\varepsilon_{\parallel}$ is the relative deformation in the loading direction.

## Poisson's ratio for an incompressible material

A material is incompressible if its volume change is zero. Mathematically, this means that the volume strain $\varepsilon_V$ is zero:

$$
\varepsilon_V = \frac{\Delta V}{V} = 0
$$

The volume strain is given by the sum of the principal normal strains:

$$
\varepsilon_V = \frac{l_x l_y l_z - l_{x0}l_{y0}l_{z0}}{l_{x0}l_{y0}l_{z0}} = \frac{(l_{x0} + \Delta l_x) (l_{y0} + \Delta l_y) (l_{z0} + \Delta l_z) - l_{x0}l_{y0}l_{z0}}{l_{x0}l_{y0}l_{z0}} = \varepsilon_x + \varepsilon_y + \varepsilon_z
$$

For an incompressible material, the following holds:

$$
\varepsilon_x + \varepsilon_y + \varepsilon_z = 0
$$

Consider a tensile load in the direction of the $x$ axis, i.e. $\varepsilon_x = \varepsilon_{\parallel}$. Due to symmetry, the transverse deformations in the $y$ and $z$ axes are equal to $\varepsilon_y = \varepsilon_z = \varepsilon_{\perp}$.

Substitute into the incompressibility condition:

$$
\varepsilon_x + 2\varepsilon_{\perp} = 0
$$

Express $\varepsilon_{\perp}$:

$$
\varepsilon_{\perp} = -\frac{\varepsilon_x}{2}
$$

Substitute into the definition of Poisson's ratio:

$$
\nu = -\frac{\varepsilon_{\perp}}{\varepsilon_x} = -\frac{-\varepsilon_x / 2}{\varepsilon_x} = \frac{1}{2}
$$

For an incompressible material:

$$
\nu = 0.5
$$

## Typical values

| Material | Poisson's ratio $\nu$ |
|-------------------|------------------------------|
| Metals | 0.25 - 0.35 |
| Steel | 0.3 |
| Aluminum | 0.33 |
| Glass | 0.2 - 0.3 |
| Concrete | 0.1 - 0.2 |
| Wood (high humidity) | 0.3 |
| Wood (dry) | 0.2 |
| Plastics (typical) | 0.3 - 0.4 |
| Rubber (elastomers) | 0.49 - 0.5 |
| Ceramics | 0.15 - 0.25 |
| Foam | -0.1 to -0.5 (auxetic) |
| Water (high pressure) | 0.5 |

## Relationship to modulus of elasticity

Poisson's ratio affects other mechanical properties of a material. For isotropic materials, there is a relationship between Poisson's ratio $\nu$, Young's modulus $E$, and shear modulus $G$:

$$
G = \frac{E}{2(1+\nu)}
$$
