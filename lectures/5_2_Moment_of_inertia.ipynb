# Moment of inertia

The moment of inertia $I$ expresses the resistance of a body to a change in rotational motion around a given axis. It is analogous to mass in translational motion.

Mathematically, it is defined as:

$$
I = \sum_i m_i r_i^2
$$

where:
- $m_i$ is the mass of a small element of the body,
- $r_i$ is the perpendicular distance of this element from the axis of rotation.

In the case of continuous mass distribution, we switch to the integral:

$$
I = \int r^2 \, dm
$$

where $dm$ is an infinitesimal mass element of the body.

## Derivation of the moment of inertia for a thin rod rotating around one end

Let us choose a thin homogeneous rod of length $L$ and mass $M$ that rotates around one end.

1. We choose the differential element $dm$ as a small piece of the rod of length $dx$.
2. The mass density (linear density) is:

$$
\lambda = \frac{M}{L}
$$

3. The differential mass element is:

$$
dm = \lambda \, dx = \frac{M}{L} dx
$$

4. The moment of inertia about one end is:

$$
I = \int_0^L x^2 \, dm
$$

Substitute $dm$:

$$
I = \int_0^L x^2 \frac{M}{L} dx
$$

5. Solve the integral:

$$
I = \frac{M}{L} \int_0^L x^2 dx
$$

$$
I = \frac{M}{L} \left[ \frac{x^3}{3} \right]_0^L
$$

$$
I = \frac{1}{3} M L^2
$$

## Table of typical moments of inertia

| Body | Axis of rotation | Moment of inertia $ I $ |
|-------|-----------|--------------------------|
| Point particle | at distance $ r $ from axis | $ m r^2 $ |
| Thin rod | around center | $ \frac{1}{12} M L^2 $ |
| Thin rod | around end | $ \frac{1}{3} M L^2 $ |
| Rectangular plate | around center, axis parallel to longer side $ a $ | $ \frac{1}{12} M a^2 $ |
| Rectangular plate | around center, axis parallel to shorter side $ b $ | $ \frac{1}{12} M b^2 $ |
| Solid cylinder | around central axis | $ \frac{1}{2} M R^2 $ |
| Solid cylinder | around axis in base plane | $ \frac{1}{4} M R^2 + \frac{1}{12} M H^2 $ |
| Hollow cylinder (thin-walled) | around central axis | $ M R^2 $ |
| Solid sphere | around diameter | $ \frac{2}{5} M R^2 $ |
| Hollow sphere (thin-walled) | around diameter | $ \frac{2}{3} M R^2 $ |
