## In-plane shear stress

Shear stress $\tau$ describes the internal forces in a material acting parallel to a given surface. It is defined as the force acting per unit area:

$$
\tau = \frac{F}{A}
$$

where:
- $F$ is the shear force acting in the plane,
- $A$ is the area over which the force acts.

Shear stress causes deformation, which we express in terms of **shear strain** ($\gamma$), which is the angle of change between originally perpendicular lines in the material.

![](https://uploads-cdn.omnicalculator.com/images/shear_strain/1.png)

Shear strain can also be expressed in terms of relative stress

$$\gamma = \frac{\Delta x}{h}$$

Shear stress is linearly proportional to shear strain, which is expressed by **Hooke's law for shear**:

$$
\tau = G \gamma
$$

where $G$ is the **shear modulus** (shear modulus), which characterizes the material's resistance to shear deformation.

## Law of combined shear stresses

> The law of combined shear stresses states that **shear stresses on two mutually perpendicular planes are equal**.

Mathematically:

$$
\tau_{xy} = \tau_{yx}
$$

### Derivation

Consider a small elementary cuboid in equilibrium. For shear forces acting on opposite faces of a cuboid, the moment equilibrium must hold. For example, for shear stresses $\tau_{xy}$ and $\tau_{yx}$:

- The force on wall $x$ in the $y$ direction induces a moment about the $z$ axis.
- The same magnitude of the moment must be induced by the force on wall $y$ in the $x$ direction.
- Therefore, $\tau_{xy} = \tau_{yx}$ must hold.

A similar procedure can be applied to other shear stresses.

## Relationship between the modulus of elasticity $G$ and Young's modulus $E$

For isotropic materials, there is a relationship between the shear modulus $G$, the Young's modulus $E$, and the Poisson's ratio $\nu$. We start from the equation for volumetric strain in Hooke's law:

$$
G = \frac{E}{2(1+\nu)}
$$

### Pure shear stress

The basic relationship between stress and strain for isotropic materials in Hooke's law in two-dimensional space is:

$$
\varepsilon_x = \frac{1}{E} \left[ \sigma_x - \nu (\sigma_y) \right]
$$

Similarly for shear strains:

$$
\gamma_{xy} = \frac{\tau_{xy}}{G}
$$

![](https://www.doitpoms.ac.uk/tlplib/mechanical_properties/images/mm_cube_stress_strain.jpg)

Consider a cube with side $a$. Assume the action of a shear stress $\tau$, which will cause a small deformation $\gamma$. The deformation of the diagonal of the cube is

$$\frac{{{\rm{\Delta }}d}}{{\sqrt 2 a}} = \frac{{\gamma a\;{\rm{cos}}45^\circ }}{{\sqrt 2 a}} = \frac{\gamma }{2} $$

From the definition of the shear modulus of elasticity, the deformation in the direction of the diagonal is

$$\epsilon_\parallel = \frac{\tau}{2G} $$

The deformation in the direction of the other diagonal is

$$\epsilon_\perp = -\frac{\tau}{2G} $$

Substituting into Hooke's law, we obtain

$$\frac{\tau}{2G} = \frac{1}{E} \left[ \sigma_\parallel - \nu \sigma_\perp \right] $$
$$-\frac{\tau}{2G} = \frac{1}{E} \left[ \sigma_\perp - \nu \sigma_\parallel \right] $$

From the above, it follows that

$$ \sigma_\perp = - \sigma_\parallel $$

and the stress can be expressed in the form.

![](https://www.doitpoms.ac.uk/tlplib/mechanical_properties/images/mm_stress_state.jpg)

From Mohr's circle, it follows that for simple shear,

$\sigma_1 =\sigma_\perp = \tau; \sigma_2 = \sigma_\parallel = -\tau;$

Therefore, we can write

$$\frac{\tau}{2G} = \frac{\tau}{E} (1 + \nu)$$

and

$$
G = \frac{E}{2(1+\nu)}
$$

## Conclusion

- Shear stress describes forces acting parallel to the surface of a body and is proportional to the shear deformation.
- The shear modulus \( G \) is related to the Young's modulus \( E \) and the Poisson's ratio \( \nu \) by the relationship:

$$
G = \frac{E}{2(1+\nu)}
$$

- The law of combined shear stresses states that the shear stresses on mutually perpendicular surfaces are equal:

$$
\tau_{xy} = \tau_{yx}, \quad \tau_{xz} = \tau_{zx}, \quad \tau_{yz} = \tau_{zy}
$$

This law follows from the equilibrium of moments in an elementary volume.
