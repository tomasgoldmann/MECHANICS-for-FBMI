### Equation of Motion

To derive the equation of motion for damped vibrations, we consider an oscillator with mass $m$, spring stiffness $k$, and a damping force proportional to velocity, $F_d$ = $-bv$, where $b$ is the damping coefficient.

According to Newton's second law:

$$\sum F = m \ddot{x}$$

where:
* $\sum F$ is the resultant force acting on the oscillator
* $m$ is the mass of the oscillator
* $a$ is the acceleration of the oscillator

<div class='center-align'><figure><img src="https://beltoforion.de/en/harmonic_oscillator/images/force1.svg" width="200" alt="Body in plane"/></figure></div>

The resultant force is the sum of the spring restoring force and the damping force:

$$\sum F = -kx - b \dot{x}$$

Substituting into Newton's second law, we get:

$$m \ddot{x} = -kx - b \dot{x}$$

$$m \ddot{x} + b \dot{x} + kx = 0$$

$$
\ddot{x} + \frac{b}{m} \dot{x} + \frac{k}{m} x = 0
$$

This equation is a linear, homogeneous second-order differential equation with constant coefficients. Determining the type of differential equation is important because the approach needed to solve it depends on it. The exponential ansatz is usually chosen to solve differential equations of this type.

## Exponential solution

The equation is a homogeneous linear second-order differential equation with constant coefficients. It is assumed that the solution to such a differential equation is an exponential function. This method was first described by Leonard Euler.

$$ x(t) = C e^{\lambda t}$$

Taking the first and second derivatives, we get:

$$
\dot{x}(t) = \lambda C e^{\lambda t}
$$

$$
\ddot{x}(t) = \lambda^2 C e^{\lambda t}
$$

Substituting these equations into the differential equation, we get:

$$
\lambda^2 C e^{\lambda t} + \frac{b}{m} \lambda C e^{\lambda t} + \frac{k}{m} C e^{\lambda t} = 0
$$

Dividing both sides by $C e^{\lambda t}$ simplifies to:

$$
\lambda^2 + \frac{b}{m} \lambda + \frac{k}{m} = 0
$$

This equation is called the **characteristic equation**. This is a quadratic equation in standard form, so there are two solutions for $\lambda$:

$$
\lambda_{1,2} = - \frac{b}{2m} \pm \sqrt{\left( \frac{b}{2m} \right)^2 - \frac{k}{m}}
$$

For further simplification, we introduce new constants $\delta$ and $\omega$:

$$
\delta = \frac{b}{2m}, \quad \omega_0 = \sqrt{\frac{k}{m}}
$$

The equation is then rewritten as:

$$
\lambda_{1,2} = - \delta \pm \sqrt{\delta^2 - \omega_0^2}
$$

The next solution depends on the values ​​of $\lambda_1$ and $\lambda_2$. Several different possibilities must be considered. The term inside the square root is called the discriminant. Depending on the choice of the constants $\delta$ and $\omega_0$, the discriminant can be greater than 0, less than 0, or equal to 0. Therefore, $\lambda_1$ and $\lambda_2$ can be:

- Two different real solutions.

- Two complex conjugate solutions.

- Two identical real solutions.

Each of these options requires a different approach to the solution. The general solution of a homogeneous differential equation has the form:

$$
x(t) = C_1 x_1(t) + C_2 x_2(t)
$$

The functions $x_1(t)$ and $x_2(t)$ are determined by the value of the discriminant. Let's now look at the different options.

## Overdamped case - aperiodic oscillation

If $\delta > \omega_0$, we have the case of strong friction. The discriminant in the equation is positive and there are two different real solutions for $\lambda$. The solution to the differential equation then looks like this:

$$
x_1(t) = C_1 e^{\lambda_1 t}
$$
$$
x_2(t) = C_2 e^{\lambda_2 t}
$$

Substituting into the equation, we get the general solution to the differential equation:

$$
x(t) = C_1 e^{(-\delta + \sqrt{\delta^2 - \omega_0^2}) t} + C_2 e^{(-\delta - \sqrt{\delta^2 - \omega_0^2}) t}
$$

When simplifying the equation, we can insert a new constant term:

$$
\alpha = \sqrt{\delta^2 - \omega_0^2}
$$

Substituting into the equation, we get:

$$
x(t) = e^{-\delta t} \left( C_1 e^{\alpha t} + C_2 e^{-\alpha t} \right)
$$

This is the solution for the case of strong damping. The constants $C_1$ and $C_2$ for a particular problem can be determined from initial conditions, such as the initial position and initial velocity of the pendulum.

$$
x(0) = x_0, \quad \dot{x}(0) = v_0
$$

## Critical case of damping - limit aperiodic oscillation

We are talking about critical damping when $\delta = \omega_0$. This is a transition from excessive damping to oscillation. In this case, the equation has a single solution for $\lambda$:

$$
\lambda_1 = \lambda_2 = \lambda = -\delta $$
It might seem that in this case we have only one solution:

$$x_1(t) = C_1 e^{\lambda t} $$

However, the given solution is not physically correct. According to it, at time $t=0$ and with the initial deflection $x_0$ the velocity is always equal to $v_0 = v(0) = x_0 \lambda$. In essence, however, at time 0 we can place the oscillator in any position $x_0$ with any velocity $v_0$. We can obtain the solution if we consider the strongly damped case when $\alpha \rightarrow 0$. For small values ​​of $\alpha$ we can write
$e^{\alpha t} = 1 + \alpha t $
and at the same time the relation
$$
x(t) = e^{-\delta t} \left( C_1 e^{\alpha t} + C_2 e^{-\alpha t} \right)
$$
is valid for all $C_1$ and $C_2$. If we choose $C_2 = -C_1$ we get
$$
x(t) = e^{-\delta t} \left( C_1 2 \alpha t \right) = t e^{\lambda t} (2 \alpha C_1)
$$
Since multiplying by a constant only changes the magnitude of the unknown constant $C_1$ we can write for the second solution
Using the exponential solution in this case means using the following procedure to obtain two partial solutions of the differential equation:

$$
x_2(t) = t C_2 e^{\lambda t}
$$

Putting into the general solution gives us the solution for the critical damping case:

$$
x(t) = e^{-\delta t} (C_1 + t C_2)
$$

The integration constants $C_1$ and $C_2$ must be obtained from the initial conditions of the particular problem:

$$
x(0) = x_0 \quad,\quad \dot{x}(0) = v_0
$$

Details for calculating the constants can be found [at Wolfram Alpha](https://www.wolframalpha.com/input/?i=solve+exp(-delta*t)*(c_1%2Bc_2*t)%3Dx_0,+-exp(-delta*t)*(delta*c_1%2Bdelta*c_2*t-c_2)+%3Dv_0,+t%3D0+for+c_1,c_2).

## Underdamped Harmonic Oscillation

Underdamped oscillation occurs for $\delta < \omega_0$. In this case, the discriminant in the equation is negative. Therefore $\lambda_1$ and $lambda_2$ are complex numbers. The exponential function $(x(t)=C e^{\lambda t}$ is a solution to the differential equation.

$$
x_1(t) = C_1 e^{\lambda_1 t}
$$

$$
x_2(t) = C_2 e^{\lambda_2 t}
$$

The expression for $\lambda is:

$$
x(t) = e^{-\delta t} \left( C_1 e^{\sqrt{\delta^2 - \omega_0^2}\; t} + C_2 e^{-\sqrt{\delta^2 - \omega_0^2}\; t}\right)
$$

The given solution expresses kmoplex numbers. because in the damped case the term under the square root is negative.
When we replace the expression under the parentheses with a new quantity $\omega_d$, we obtain

$$
\omega_d = \sqrt{\omega_0^2 - \delta^2}
$$

$$
\sqrt{\delta^2 - \omega_0^2} = \sqrt{-1 \cdot (\omega_0^2 - \delta) } = i \sqrt{\omega_0^2 - \delta^2} \tag{eq:iomega}
$$

As we will see in a moment, the constant $\omega_d$ represents the natural frequency of the damped harmonic oscillator. We can express the solution to the equation of motion as

$$
x(t) = e^{-\delta t} \left( {C_1 e^{i \omega_d t} + C_2 e^{- i \omega_d t}} \right)
$$

To work with complex numbers, we use Euler's formula, which connects complex exponential functions with trigonometric functions:

$$
e^{i \phi} = \cos \phi + i \sin \phi
$$

$$
x(t) = e^{-\delta t} \left( C_1 \cos \omega_d t + i C_1 \sin \omega_d t + C_2 \cos_d \omega t - i C_2 \sin \omega_d t\right)
$$

$$
x(t) = e^{-\delta t} \left( (C_1 + C_2) \cos \omega_d t + i (C_1-C_2) \sin \omega_d t \right)
$$

$$
x(t) = e^{-\delta t} \left( C_3 \cos \omega_d t + C_4 \sin \omega_d t \right)
$$
where $C_4$ is $i (C_1-C_2)$. We have shown that the sum of a pair of harmonic oscillations with the same frequency can be expressed as a single trigonometric function that includes the phase shift

$$
x(t) = e^{-\delta t} \left( A \cos \omega_d t + \psi \right)
$$

where
* $\omega_d$ is the damped natural frequency
* the constants $A$ and $\psi$ are the amplitude and phase and can be calculated from a given set of initial conditions.

$$
\omega_d = \sqrt{\omega_0^2 - \delta^2} = \sqrt{\omega_0^2 - \left(\frac{b}{2m}\right)^2} = \omega_0 \sqrt{1 - \left(\frac{b}{2m \omega_0}\right)^2}
$$

The period of the damped oscillations is determined

$$
T_d = \frac{2\pi}{\omega_d}
$$

and the frequency

$$
f_d = \frac{\omega_d}{2\pi} = f \sqrt{1 - \left(\frac{b}{2m \omega_0}\right)^2}
$$

where $f$ is the frequency of the natural oscillations

# Damping coefficient

We can introduce a quantity that we will denote as **damping coefficient**

$$ \zeta = \frac{b}{2m\omega_0}$$

In the notation we used in the previous text, the following applies:

$$\sqrt{\delta^2 - \omega_0^2} = \omega_0 \sqrt{\zeta^2 - 1} $$

and therefore we can judge the type of oscillations based on the value of the damping coefficient

1. $\zeta > 1$ - aperiodic oscillation
2. $\zeta = 1$ - limiting aperiodic oscillation
3. $\zeta < 0$ - damped oscillation

## Visualization

To plot the response, we will use the solution to the differential equation we derived. The solution for a damped harmonic is:

$$ x(t) = e^{-\zeta\omega_0t}\left(c_1 e^{i \omega_d t} + c_2 e ^{-i \omega_d t}\right) $$

$$ x(t) = e^{-\zeta\omega_0t}\left(a_1 \cos{\omega_d t} + a_2 \sin{\omega_d t}\right) $$

To use this equation, we need to solve for $c_1$ and $c_2$ or $a_1$ and $a_2$ using initial conditions. Here, we will use the sin/cos form. Solving the equation for the general initial velocity $\\dot{x} = v_0$ and the general initial deflection $x = x_0$ we find:

$$ x(t) = e^{-\zeta \omega_0 t}\left(x_0 \cos{\omega_d t} + \frac{\zeta \omega_n x_0 + v_0}{\omega_d} \sin{\omega_d t}\right)$$

```python
# Define the System Parameters
m = 1.0                   # kg
k = (2.0 * np.pi)**2.     # N/m (Selected to give an undamped natrual frequency of 1Hz)
wn = np.sqrt(k/m)         # Natural Frequency (rad/s)

z = 0.1                   # Define a desired damping ratio
c = 2*z*wn*m              # calculate the damping coeff. to create it (N/(m/s))

wd = wn*np.sqrt(1 - z**2) # Damped natural frequency (rad/s)
# Set up simulation parameters
t = np.linspace(0, 5, 501)          # Time for simulation, 0-5s with 501 points in-between

# Define the initial conditions x(0) = 1 and x_dot(0) = 0
x0 = np.array([-1.0, 0.0])
# Define x(t)
x = np.exp(-z*wn*t)*(x0[0]*np.cos(wd*t) + (z*wn*x0[0] + x0[1])/wd * np.sin(wd*t))
# # Make the figure pretty, then plot the results
# #   "pretty" parameters selected based on pdf output, not screen output
# #   Many of these setting could also be made default by the .matplotlibrc file

# Set the plot size - 3x2 aspect ratio is best
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
plt.setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
plt.xlabel('Time (s)', family='serif', fontsize=22, weight='bold', labelpad=5)
plt.ylabel('Distance', family='serif', fontsize=22, weight='bold', labelpad=10)


amp = np.sqrt(x0[0]**2 + ((z*wn*x0[0] + x0[1])/wd)**2)
decay_env = amp * np.exp(-z*wn*t)

# plot the decay envelope
plt.plot(t, decay_env, linewidth=1.0, linestyle = '--', color = "red")
plt.plot(t, -decay_env, linewidth=1.0, linestyle = '--', color = "red")

plt.plot(t, x, linewidth=2, linestyle = '-', label=r'Response')

# uncomment below and set limits if needed
# xlim(0,5)
# ylim(0,10)

plt.yticks([-1.5, -1, -0.5, 0, 0.5, 1, 1.5], ['', r'$-x_0$', '', '0', '', r'$x_0$', ''])

plt.annotate('Exponencial envelope',
         xy=(t[int(len(t)/3)],decay_env[int(len(t)/3)]), xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=18,
         arrowprops=dict(arrowstyle="simple, head_width = 0.35, tail_width=0.05", connectionstyle="arc3, rad=.2", color="#377eb8"), color = "#377eb8")


# # Create the legend, then fix the fontsize
# leg = plt.legend(loc='upper right', fancybox=True)
# ltext  = leg.get_texts()
# plt.setp(ltext,family='serif',fontsize=18)

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad = 0.5)

# save the figure as a high-res pdf in the current folder
# It's saved at the original 6x4 size
# plt.savefig('MCHE485_FreeVibrationWithDamping.pdf')

fig.set_size_inches(9, 6) # Resize the figure for better display in the notebook
```
