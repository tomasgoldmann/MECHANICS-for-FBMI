# Dynamics of rotational motion

Newton's second law in vector form states that the resultant force acting on a body is equal to the change in momentum with time:

$$
\vec{F} = \frac{d\vec{p}}{dt}
$$

where $\vec{p} = m \vec{v}$ is the momentum. We now apply this equation to rotational motion.

## Moment of force and moment of momentum

For rotational motion, we introduce the following quantities:

- **Torque:**
The moment of force relative to point $O$ is defined as:

$$
\vec{M}_O = \vec{r} \times \vec{F}
$$

where $\vec{r}$ is the position vector of the mass point relative to point $O$, $\vec{F}$ is the force acting on the body, and $\times$ denotes the vector product.

- **Momentum of Momentum:**
The moment of moment of a mass point relative to point $O$ is defined as:

$$
\vec{L}_O = \vec{r} \times \vec{p} = \vec{r} \times (m \vec{v})
$$

For a system of multiple particles, the total moment of moment is given by the sum of the momenta of the individual particles:

$$
\vec{L}_O = \sum_i \vec{r}_i \times m_i \vec{v}_i
$$

## 2. Derivation of Momentum

We apply the time derivative to the moment of moment:

$$
\frac{d\vec{L}_O}{dt} = \frac{d}{dt} \sum_i (\vec{r}_i \times m_i \vec{v}_i)
$$

We use the product derivative rule:

$$
\frac{d\vec{L}_O}{dt} = \sum_i \left( \frac{d\vec{r}_i}{dt} \times m_i \vec{v}_i + \vec{r}_i \times m_i \frac{d\vec{v}_i}{dt} \right)
$$

Since $\frac{d\vec{r}_i}{dt} = \vec{v}_i$, we have:

$$
\vec{v}_i \times \vec{v}_i = 0
$$

so the first term vanishes and we are left with:

$$
\frac{d\vec{L}_O}{dt} = \sum_i \vec{r}_i \times m_i \frac{d\vec{v}_i}{dt}
$$

From Newton's second law $m_i \frac{d\vec{v}_i}{dt} = \vec{F}_i$ we get:

$$
\frac{d\vec{L}_O}{dt} = \sum_i \vec{r}_i \times \vec{F}_i
$$

The term on the right-hand side is the resultant moment of force acting on the body:

$$
\frac{d\vec{L}_O}{dt} = \vec{M}_O
$$

which is **the general equation of rotational motion**.

## 3. Special case â€“ rotational motion around a fixed axis

If a body rotates around a fixed axis, the angular momentum can be expressed as:

$$
\vec{L} = I \vec{\omega}
$$

where:
- $I$ is the moment of inertia about the given axis,
- $\vec{\omega}$ is the angular velocity.

We substitute into the basic equation:

$$
\vec{M} = \frac{d}{dt} (I \vec{\omega})
$$

If the moment of inertia is constant, we can write \( I \):

$$
\vec{M} = I \frac{d\vec{\omega}}{dt} = I \vec{\alpha}
$$

where $\vec{\alpha}$ is the angular acceleration.
