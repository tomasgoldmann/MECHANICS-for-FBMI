# Engineering vs. true stress and strain

```python
import numpy as np
import matplotlib.pyplot as plt
```

In engineering mechanics and strength analysis, **enginnering stress** and **enginnering strain** are used to describe material behavior. These quantities are introduced mainly in connection with tensile tests and material property analysis.

![](https://yasincapar.com/wp-content/uploads/2020/10/Untitled-Project2.gif)

## Engineering stress

> Contractual stress $\sigma_s$ is defined as the ratio of the current force $F$ to the **initial cross-sectional area** $A_0$ before deformation:

$$
\sigma_s = \frac{F}{A_0}
$$

where:
- $F$ is the applied force \[N}\],
- $A_0$ is the initial cross-sectional area before deformation \[m\].

This approach simplifies calculations, because with large plastic deformations, the cross-section changes significantly, which would complicate the calculation of the true stress.

### Difference between enginnering and true stress
- **Enginnering stress** – takes into account the original cross-section $A_0$, which is common in engineering calculations.
- **True stress** – is based on the current cross-sectional area $S$ at a given moment of deformation, which is more accurate, but more complicated to calculate:

$$
\sigma_t = \frac{F}{A}
$$

where $A$ is the actual cross-sectional area at the moment of measurement.

## Engineering strain

> Engineering strain $(\varepsilon_s)$ is defined as the ratio of the elongation of the length to the original length of the sample:

$$
\varepsilon_s = \frac{\Delta L}{L_0}
$$

where:
- $\Delta L = L - L_0$ is the change in length \[m\],
- $L_0$ is the original length of the sample \[m\]).
-

![](https://yasincapar.com/wp-content/uploads/2020/10/11-2.png)

### Difference between enginnering and true strain
- **Enginnering strain** – uses the initial length $L_0$, which simplifies calculations.
- **True (logarithmic) strain** – takes into account the instantaneous length \(L\), which is more accurate especially for large deformations

## Derivation of the logarithmic strain relation

We start from the definition of infinitesimal increment of longitudinal strain:

$$
d\varepsilon_t = \frac{dL}{L}
$$

Integrating this equation in the limits from $L_0$ (initial length) to $L$ (current length) we obtain:

$$
\varepsilon_t = \int_{L_0}^{L} \frac{dL}{L}
$$

This integral is elementary and **logarithmic strain** is given by the relation:

$$
\varepsilon_t = \ln{\left( \frac{L}{L_0} \right)}
$$

### Relationship between logarithmic and enginnering strain

For small deformations ($\varepsilon_s \ll 1$) the logarithmic strain is approximately equal to the enginnering strain:

$$
\ln(1 + \varepsilon_s) \approx \varepsilon_s, \quad \text{for } \varepsilon_s \ll 1
$$

However, for large strains:

$$
\varepsilon_t = \ln{\frac{L}{L_0}} = \ln\left({\frac{L0}{L_0} + \frac{L-L0}{L_0}}\right) = \ln(1 + \varepsilon_s)
$$

> **Caution**
>
>The logarithmic strain is always slightly smaller than the engineering strain for larger values ​​of $\varepsilon_s$.

```python
L0 = 1
delta_L = 0.5
L = np.linspace(L0,L0 + delta_L,100)

epsilon_s = (L - L0)/L0
epsilon_t = np.log(L/L0)
# for numpy log denotes natural logarithm

plt.plot(L, epsilon_s, label = "engineering strain")
plt.plot(L, epsilon_t, label = "true strain")
plt.xlabel("Sample length (m)")
plt.ylabel("Ratio strain [1]")
plt.legend()
plt.show()
```
