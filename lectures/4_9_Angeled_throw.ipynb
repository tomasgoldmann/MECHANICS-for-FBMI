# Derivation of the angeled throw from the Lagrange-Euler equations

We derive the equations of the angeled throw using the Lagrange-Euler equations.

Consider a projectile of mass $m$ moving in a 2D plane (x, y) under the influence of gravity $g$. The Lagrangian approach uses generalized coordinates, which in this case are simply the Cartesian coordinates *x* and *y*.

* **Kinetic energy (Ek):** The kinetic energy of the projectile is given by:

$$E_k = \frac{1}{2}m(\dot{x}^2 + \dot{y}^2)$$

where $\dot{x}$ and $\dot{y}$ are the time derivatives of *x* and *y* (i.e., the velocities in the x and y directions).
* **Potential Energy (U):** The potential energy of a projectile due to gravity is given by:

$$U = mgy$$

where *y* is the vertical coordinate.

**Definition of the Lagrangian (L):**

* The Lagrangian is defined as the difference between kinetic and potential energy:

$$L = E_k - U = \frac{1}{2}m(\dot{x}^2 + \dot{y}^2) - mgy$$

**Application of the Lagrange-Euler equations:**

* The Lagrange-Euler equations are given by:

$$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{q}_i}\right) - \frac{\partial L}{\partial q_i} = 0$$

where $q_i$ are the generalized coordinates (in this case *x* and *y*).

**Derivation of the equations of motion for x:**

* For the coordinate *x* we have:

$$\frac{\partial L}{\partial \dot{x}} = m\dot{x}$$

$$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{x}}\right) = m\ddot{x}$$

$$\frac{\partial L}{\partial x} = 0$$

Substituting these values ​​into the Lagrange-Euler equation:

$$m\ddot{x} - 0 = 0$$

$$\ddot{x} = 0$$

This equation means that the acceleration in the x direction is zero, which implies a constant velocity.

**Deriving the equations of motion for y:**

* For the *y* coordinate we have:

$$\frac{\partial L}{\partial \dot{y}} = m\dot{y}$$

$$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{y}}\right) = m\ddot{y}$$

$$\frac{\partial L}{\partial y} = -mg$$

Substituting these values ​​into the Lagrange-Euler equation:

$$m\ddot{y} - (-mg) = 0$$

$$m\ddot{y} + mg = 0$$

$$\ddot{y} = -g$$

This equation means that the acceleration in the y direction is -g, which is the acceleration due to gravity.

**Solving differential equations:**

* From $\ddot{x} = 0$ we get:

$$\dot{x} = v_{0x}$$

$$x = v_{0x}t + x_0$$

where $v_{0x}$ is the initial velocity in the x direction and $x_0$ is the initial x coordinate.
* From $\ddot{y} = -g$ we get:

$$\dot{y} = -gt + v_{0y}$$

$$y = -\frac{1}{2}gt^2 + v_{0y}t + y_0$$

where $v_{0y}$ is the initial velocity in the y direction and $y_0$ is the initial y coordinate.

**Conclusion:**

The derived equations:

* $x = v_{0x}t + x_0$
* $y = -\frac{1}{2}gt^2 + v_{0y}t + y_0$

are the standard equations of motion for an oblique throw, showing that the horizontal motion is uniform and the vertical motion is uniformly accelerated by gravity. This shows how the Lagrange-Euler equations can be used to derive the equations of motion for a simple projectile.
