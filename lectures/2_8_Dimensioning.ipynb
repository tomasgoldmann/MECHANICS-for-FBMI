# Dimensioning of components under tension and compression

The dimensioning of components under tension and compression is carried out so that the component is able to safely transmit a given load without plastic deformation or fracture. The main criteria are **mechanical strength of the material**, **safety factor** and **limit states of the component**.

## Dimensioning principles

Two basic methods are used when dimensioning components:

1. **Load dimensioning** – we determine the maximum permissible force that the component can transmit.
2. **Cross-section dimensioning** – we design the corresponding cross-section of the component so that at a given force the stress does not exceed the permissible value.

### **Load dimensioning**

Force dimensioning consists in determining the maximum permissible force $F_{\text{max}}$ that the component can withstand without failure.

From the basic equation for normal stress:

$$
\sigma = \frac{N}{A}
$$

where:
- $\sigma$ is the normal stress \[Pa\],
- $N$ is the applied internal force \[N\],
- $A$ is the cross-sectional area of ​​the component \[m²\].

The maximum allowable force is given by:

$$
N_{\text{max}} = \sigma_\text{dov} A
$$

where $\sigma_{\text{dov}}$ is the allowable stress of the material, which depends on the strength of the material and the safety factor.

> **Caution**
>
>In the case of variable load and cross-section along the length of the member, all parts must satisfy the strength condition.

### **Sizing by cross-section**

When sizing by cross-section, we determine the required cross-sectional area $A_{\text{min}}$, which ensures that the allowable stress is not exceeded when a given force is applied:

$$
A_{\text{min}} = \frac{N}{\sigma_{\text{dov}}}
$$

The selected cross-section must be large enough for the component to be safely stressed.

## Safety factor

> The safety factor $K$ is the ratio between the ultimate strength of the material and the actual or allowable stress:
>
> $$K = \frac{\sigma_{\text{mez}}}{\sigma_{\text{dov}}}$$
>

where:
- $\sigma_{\text{mez}}$ is the strength limit of the material (yield limit or ultimate strength),
- $\sigma_{\text{dov}}$ is the allowable stress at which the component should not fail.

The safety factor takes into account uncertainties in the design, for example:
- manufacturing tolerances,
- variability of material properties,
- dynamic loading and material fatigue,
- safety requirements in service.

The allowable stress is determined according to the relationship:

$$
\sigma_{\text{dov}} = \frac{\sigma_{\text{mez}}}{K}
$$

> **Note**
>
>The higher the safety factor, the more conservative the design, but at the same time leads to more robust and heavier components.
