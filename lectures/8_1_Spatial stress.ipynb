# Spatial stress

## Definition of stress

Stress describes the state of internal stress in a material caused by external loads. It is characterized by **stress**, which is the force per unit area. The stress vector on a particular surface in a body depends on the orientation of that surface and the point in the body.

## Principles of spatial stress

Consider a square-sectioned bar subjected to a tensile force **F**. This force causes a stress in the bar

![](https://eng.libretexts.org/@api/deki/files/19903/2.1.1.png?revision=1)

$$
T = \sigma = \frac{\text{force}}{\text{area}} = \frac{F}{A_0} \quad [\text{N}/\text{mm}^2]
$$

where $\vec{T}$ is also defined as the tension vector or traction vector

$$
\vec{T} = \frac{\vec{F}}{A_0}
$$

In the uniaxial case, the surface tension is the only component of the stress tensor in the global coordinate system, commonly denoted as $\sigma$.

>note
>
> How to apply force to the end of a rod?
This can be done in several different ways (see image below). A pin connector can be glued (or welded) to the end of the rod, or a hole can be drilled through the rod for the pin to attach.

![Methods of applying tensile force](https://eng.libretexts.org/@api/deki/files/19909/2.1.2.png?revision=1)

Alternatively, an internal or external thread can be produced. Finally, the axial force can be applied using friction or mechanical jaws.

Except in cases with welded or glued joints, a complex state of stress arises in the region of the ends of the member, which is multicomponent (multiaxial). Such a state of stress is limited to a relatively short section of the member, comparable to its height or diameter. In this section, there is a gradual transition from the multicomponent state of stress to the uniaxial state, for which equation (2.1.1) applies.

The above example can serve as a practical application of **Saint-Venant's principle** (1856). This principle, named after the French elasticity theorist **Jean-Claude Barrém de Saint-Venant**, can be formulated as follows:

> "The difference between the effects of two different but statically equivalent loads becomes very small at a sufficiently large distance from the point of load."


Let's think about it - what are the "two" statically equivalent loads applied to the ends of the member? Usually we consider a cross-section perpendicular to the axis of the member. But now let's consider two sections at angles $\theta$ and $\pi/2-\theta$ to the normal direction. These planes are defined by the unit normal vector $\vec{n}$.

![Figure 2.1.3: Forces on an oblique section](https://eng.libretexts.org/@api/deki/files/19910/2.1.3.png?revision=1)

From the release of the body we get the components of the normal and tangential force:

$$
F_N = F \cos\theta
$$

$$
F_n = F \cos\left(\frac{\pi}{2} - \theta\right)
$$

$$
F_T = F \sin\theta
$$

$$
F_t = F \sin\left(\frac{\pi}{2} - \theta\right)
$$

The inclined cross-section **A** is larger and is related to the reference cross-section as follows:

$$
A_0 = A_A \cos\theta,\quad A_0 = A_B \cos\left(\frac{\pi}{2} - \theta\right)
$$

Now consider a unit volume cubic element located at the intersection of sections AA and BB.

A volume element with surface tension acting on two adjacent walls.

![Figure 2.1.4: Surface tension on adjacent surfaces](https://eng.libretexts.org/@api/deki/files/19904/2.1.4.png?revision=1)


The surface tension (force per unit area) on two perpendicular planes is:

- Plane parallel to AA:
$$ T_N = T \cos^2 \theta $$
$$ T_T = T \sin \theta \cos \theta $$

- Plane parallel to BB:
$$ T_n = T \cos^2 \left(\frac{\pi}{2}) - \theta \right) $$
$$ T_t = T \sin \left(\frac{\pi}{2}) - \theta \right) \cos \left(\frac{\pi}{2}) - \theta
\right) $$

It can be noted that the tangential components of the surface traction vector on sections AA and BB are identical. Normalized plots of the above quantities as a function of the angle of orientation of the section are shown in the figure.

![Figure 2.1.5](https://eng.libretexts.org/@api/deki/files/19905/2.1.5.png?revision=1)

It can be seen that the tangential component reaches a maximum at an angle of 45°. This means that if the material fails due to shear loading, the fracture surface will always be oriented at an angle of 45°. The above example teaches us that there are infinitely many combinations of the normal and tangential components of the surface stresses that are in equilibrium with the applied load. For each orientation of the cross-section, there is a different pair $\{T_n, T_t\}$.

The orientation of a surface element is uniquely determined by the unit normal vector $\mathbf{n} = [n_1, n_2, n_3]$. At the same time, the components of the surface stress vector (traction vector) acting on the same element are $\mathbf{T} = [T_1, T_2, T_3]$.

## Cauchy stress tensor

The state of stress of a material is therefore defined by the set of all stress vectors $\vec{T}^n$ associated with an infinite number of planes intersecting a given point. In two dimensions, the solution was Mohr's circle, where from the knowledge of two planes it was possible to calculate the stress in any other plane. In three dimensions, it is logical to assume that we will need three planes. For the sake of simplicity, let us consider the state where the individual planes are perpendicular to each other and thus form an elementary prism. The stress in any other plane can be obtained by means of a coordinate transformation.

For simplicity, if we consider a tetrahedron, where three of its sides are defined by the planes of the coordinate system and the fourth side is defonned by the unit vector $\vec{n}$ and size $\mathrm{d}A$

![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/eb/Cauchy_tetrahedron.svg/495px-Cauchy_tetrahedron.svg.png)


The stress vector in a given plane is defined as $\vec{T}_n$. Similarly, we can define the vectors of the remaining planes as $\vec{T}_i$, $\vec{T}_j$ and $\vec{T}_k$, where $\vec{i}, \vec{j}, \vec{k}$ are unit vectors (versors) in the $x, y$ and $z$ directions. This tetrahedron is sometimes referred to as the *Cauchy tetrahedron*. If we assume that the volume of the tetrahedron is small, we can neglect its weight and for the balance of forces we have

$$
\vec{T}_n dA - \vec{T}_i dA_x - \vec{T}_j dA_y - \vec{T}_k dA_z = 0
$$

Where negative signs indicate that the normal is in the negative direction of the given axis. For a given surface, the size of the surface is given by the projection onto the surface $\mathrm{d}A$ onto the given plane.

$$
\mathrm{d}A_x = (\vec{n} \cdot \vec{i}) \mathrm{d}A = n_x \mathrm{d}A
$$

$$
\mathrm{d}A_y = (\vec{n} \cdot \vec{j}) \mathrm{d}A = n_y \mathrm{d}A
$$

$$
\mathrm{d}A_z = (\vec{n} \cdot \vec{k}) \mathrm{d}A = n_z \mathrm{d}A
$$

We can thus simplify the force balance to the form

$$
\vec{T}_n = \vec{T}_i n_x + \vec{T}_j n_y - \vec{T}_k n_z = 0
$$

where we can express the individual stress vectors (traction vectors) using their components

$$\vec{T}_x = T_{xx} \vec{i} + T_{xy} \vec{j} + T_{xz} \vec{k} = \sigma_{xx} \vec{i} + \sigma_{xy} \vec{j} + \sigma_{xz} \vec{k}$$
$$\vec{T}_y = T_{yx} \vec{i} + T_{yy} \vec{j} + T_{yz} \vec{k} = \sigma_{yx} \vec{i} + \sigma_{yy} \vec{j} + \sigma_{yz} \vec{k}$$
$$\vec{T}_z = T_{zx} \vec{i} + T_{zy} \vec{j} + T_{zz} \vec{k} = \sigma_{zx} \vec{i} + \sigma_{zy} \vec{j} + \sigma_{zz} \vec{k}$$

The nine components of the stress vectors can be written in the form of a second-order tensor

$$
\boldsymbol{\sigma} = \sigma_{ij} = \begin{bmatrix}
\vec{T}_x \\
\vec{T}_y \\
\vec{T}_z
\end{bmatrix}
= \begin{bmatrix}
\sigma_{xx} & \sigma_{xy} & \sigma_{xz} \\
\sigma_{yx} & \sigma_{yy} & \tau_{yz} \\
\sigma_{zx} & \sigma_{zy} & \sigma_{zz}
\end{bmatrix} =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & \tau_{xz} \\
\tau_{yx} & \sigma_{yy} & \tau_{yz} \\
\tau_{zx} & \tau_{zy} & \sigma_{zz}
\end{bmatrix}
$$

The stress vector in a free plane can be determined by

$$
\vec{T}_n = \vec{T}_i n_x + \vec{T}_j n_y - \vec{T}_k n_z = \sum\limits_i \left(\sum\limits_j \sigma_{ij} \vec{e}_j\right) n_i =
\sum\limits_j \left(\sum\limits_i \sigma_{ij} {n}_i\right) \vec{e}_j = \sum_j T^n_j \vec{e}_j
$$

Alternatively, we can write it in matrix notation

$$
\left[T_x, T_y, T_z\right] = \left[ n_x, n_y, n_z \right] \begin{bmatrix}
\sigma_{xx} & \sigma_{xy} & \sigma_{xz} \\
\sigma_{yx} & \sigma_{yy} & \tau_{yz} \\
\sigma_{zx} & \sigma_{zy} & \sigma_{zz}
\end{bmatrix}
$$

or simply

$$\vec{T_n} = \vec{n} \cdot \boldsymbol{\sigma}$$


### Components of the tension vector (traction vector) in a given plane

The components of the surface traction vector acting on a plane are $\mathbf{T} = [T_1, T_2, T_3]$. For example, the orientation of the planes of a unit mass cube is shown in the figure.

![](https://eng.libretexts.org/@api/deki/files/19906/2.1.6.png?revision=1)


The relationship between the stress vectors in given planes, the unit normal vector defining the plane, and the stress tensor is given by the famous Cauchy formula:

$$ T_i = \sum\limits_i \sigma_{ij} n_i $$

or in expanded notation:

$$ T_1 = \sigma_{j1} n_j = \sigma_{11} n_1 + \sigma_{21} n_2 + \sigma_{31} n_3 $$

$$ T_2 = \sigma_{j1} n_j = \sigma_{12} n_1 + \sigma_{22} n_2 + \sigma_{32} n_3 $$

$$ T_3 = \sigma_{j1} n_j = \sigma_{13} n_1 + \sigma_{23} n_2 + \sigma_{33} n_3 $$

### Two-dimensional example: Oblique cut

In this case, a local coordinate system rotated with respect to the 3-axis is used. In this system, the components of $n$ are the same as in the global system. In a given slice, the components of the vector $\vec{n}$ are

$$
\vec{n} = [\cos \theta, \sin \theta, 0]
$$

Substituting the above values into Cauchy's formula, we obtain:

$$
\vec{T}_n = [\sigma_0 \cos \theta, 0, 0]
$$
where the components are expressed in the original coordinate system. When we want to express the normal and shear components of the stress, we must project the components of the stress tensor onto the individual axes, i.e. decompose the stress vector into two components $[T_N, T_T]$.

$$T_N = \vec{T}_n \cdot \vec{n}$$

$$T_T = \vec{T}_n \cdot \vec{m}$$

where $\vec{m} \perp \vec{n}$, and thus $\vec{m} = [\sin \theta, -\cos \theta, 0]$

$$T_N = \sigma_0 \cos^2 \theta$$

$$T_T = \sigma_0 \cos \theta \sin \theta $$


### Symmetry of the stress tensor

![](https://www.pasqualerobustini.com/wp-content/uploads/2014/10/geodin-plane-stress.gif)


The symmetry of the stress tensor comes from the moment equilibrium equation of an infinitely small volume element.

Let us imagine an infinitely small cuboid with sides $dx$, $dy$ and $dz$. For simplicity, let us consider the moment equilibrium about the $z$ axis. The forces acting on the walls perpendicular to the $x$ and $y$ axes generate moments about the $z$ axis.

* Moment from stress $\sigma_{yx}$: $M_y = \sigma_{yx} \, dx \, dy \, dz$
* Moment from stress $\sigma_{xy}$: $M_x = -\sigma_{xy} \, dx \, dy \, dz$

The moment equilibrium condition about the $z$ axis ($\sum M_z = 0$) gives:

$\sigma_{yx} \, dx \, dy \, dz - \sigma_{xy} \, dx \, dy \, dz = 0$

Assuming $dx \, dy \, dz \neq 0$, it follows:

$\sigma_{xy} = \sigma_{yx}$

Analogously, the symmetry can be derived for the other components of the stress tensor from the moment equilibrium around the $x$ and $y$ axes:

* $\sigma_{yz} = \sigma_{zy}$
* $\sigma_{xz} = \sigma_{zx}$

The symmetry of the stress tensor ($\sigma_{ij} = \sigma_{ji}$) is a necessary condition for the moment equilibrium of an infinitely small volume element. Failure to respect this symmetry would lead to a non-zero resultant moment and thus an infinitely large angular acceleration of the element, which is physically impossible in a static or quasi-static state.

>warning
>
>This derivation is valid assuming the absence of volume moments acting on the element. In engineering applications, this assumption is usually met.


The symmetry of the stress tensor is based on the moment equilibrium equation of an infinitely small volume element. In general:

$$\sigma_{ij} = \sigma_{ji} $$

This reduces the nine components of the $3 times 3$ matrix to just six independent components.

### Deformation tensor

The definition of the deformation tensor $\varepsilon_{ij}$ in continuum mechanics describes the local deformation of a material and is closely related to the displacement field $\mathbf{u}$, whose components are $u_i$ (where $i = 1, 2, 3$ corresponds to the axes $x_1, x_2, x_3$).

Consider an infinitesimal vector element $d\mathbf{x}$ in an undeformed solid, connecting two nearby points with coordinates $x_i$ and $x_i + dx_i$. After deformation, these points are shifted by $\mathbf{u}(x_i)$ and $\mathbf{u}(x_i + dx_i)$, and the vector element is transformed to $d\mathbf{x}'$.

The displacement of the endpoint of the element $d\mathbf{x}$ can be expressed using the Taylor expansion of the displacement field around the initial point:

$$
u_i(x_j + dx_j) = u_i(x_j) + \frac{\partial u_i}{\partial x_j} dx_j + O((dx)^2)
$$

We will neglect higher order terms because we are considering an infinitely small element.

The deformed vector element $d\mathbf{x}'$ connects the points with coordinates $x_i + u_i(x_j)$ and $x_i + dx_i + u_i(x_j + dx_j)$. It can be expressed as:

$$
dx'_i = (x_i + dx_i + u_i(x_j + dx_j)) - (x_i + u_i(x_j))
$$

$$
dx'_i = dx_i + u_i(x_j + dx_j) - u_i(x_j)
$$

By substituting the Taylor expansion for $u_i(x_j + dx_j)$:

$$
dx'_i = dx_i + \left( u_i(x_j) + \frac{\partial u_i}{\partial x_j} dx_j \right) - u_i(x_j)
$$

$$
dx'_i = dx_i + \frac{\partial u_i}{\partial x_j} dx_j
$$

This relationship can be written in tensor notation as:

$$
dx'_i = (\delta_{ij} + \frac{\partial u_i}{\partial x_j}) dx_j = F_{ij} dx_j
$$

where $F_{ij} = \delta_{ij} + \frac{\partial u_i}{\partial x_j}$ is the **deformation gradient tensor**. This tensor describes the linear transformation of the undeformed element $d\mathbf{x}$ to the deformed element $d\mathbf{x}'$.

To quantify the deformation itself, i.e. the change in shape and size, it is useful to introduce the **true deformation measure tensor (Cauchy-Green deformation tensor)** $\mathbf{C}$, defined as:

$$
\mathbf{C} = \mathbf{F}^T \mathbf{F} \implies C_{ij} = F_{ki} F_{kj} = \left( \delta_{ki} + \frac{\partial u_k}{\partial x_i} \right) \left( \delta_{kj} + \frac{\partial u_k}{\partial x_j} \right)
$$

$$
C_{ij} = \delta_{ij} + \frac{\partial u_j}{\partial x_i} + \frac{\partial u_i}{\partial x_j} + \frac{\partial u_k}{\partial x_i} \frac{\partial u_k}{\partial x_j}
$$

**The engineering strain rate tensor** $\mathbf{E}$ is defined as the difference between the deformed and undeformed metric tensor (which for Cartesian coordinates is the Kronecker delta $\delta_{ij}$):

$$
\mathbf{E} = \frac{1}{2} (\mathbf{C} - \mathbf{I}) \implies E_{ij} = \frac{1}{2} (C_{ij} - \delta_{ij})
$$

By substituting for $C_{ij}$:

$$
E_{ij} = \frac{1}{2} \left( \delta_{ij} + \frac{\partial u_j}{\partial x_i} + \frac{\partial u_i}{\partial x_j} + \frac{\partial u_k}{\partial x_i} \frac{\partial u_k}{\partial x_j} - \delta_{ij} \right)
$$

$$
E_{ij} = \frac{1}{2} \left( \frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i} + \frac{\partial u_k}{\partial x_i} \frac{\partial u_k}{\partial x_j} \right)
$$

For **small deformations**, where the displacement gradients $\frac{\partial u_k}{\partial x_i}$ are small compared to unity, the quadratic term $\frac{\partial u_k}{\partial x_i} \frac{\partial u_k}{\partial x_j}$ can be neglected. In this case, the engineering strain tensor reduces to the **small strain tensor** or **linearized strain tensor** $\varepsilon_{ij}$:

$$
\varepsilon_{ij} \approx \frac{1}{2} \left( \frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i} \right)
$$

This definition you provided is therefore a **linearized definition of the deformation tensor**, valid for small deformation cases. It represents the symmetric part of the displacement gradient. The symmetric part is related to the change in shape and size, while the antisymmetric part (rotation tensor) is related to the rotation of the element without changing its shape and size in the first order of approximation.

## Components of the deformation tensor

In a three-dimensional state of stress, the second-order symmetric deformation tensor $\boldsymbol{\varepsilon}$ in Cartesian coordinates (x, y, z or $x_1, x_2, x_3$) has the following six independent components:

**Normal deformations (diagonal components):**

* $\varepsilon_{xx} = \varepsilon_{11} = \frac{\partial u_x}{\partial x} = \frac{\partial u_1}{\partial x_1}$: Proportional change in the length of the element in the x-axis direction.
* $\varepsilon_{yy} = \varepsilon_{22} = \frac{\partial u_y}{\partial y} = \frac{\partial u_2}{\partial x_2}$: Proportional change in the length of the element in the y-axis direction.
* $\varepsilon_{zz} = \varepsilon_{33} = \frac{\partial u_z}{\partial z} = \frac{\partial u_3}{\partial x_3}$: Proportional change in the length of the element in the z-axis direction.

**Shear deformations (non-diagonal components):**

The tensor shear strains ($\varepsilon_{ij}$ for $i \neq j$) are related to the half-change in angle between originally perpendicular line elements. The engineering shear strain ($\gamma_{ij}$) is $\gamma_{ij} = 2 \varepsilon_{ij}$.

* $\varepsilon_{xy} = \varepsilon_{yx} = \varepsilon_{12} = \varepsilon_{21} = \frac{1}{2} \left( \frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x} \right) = \frac{1}{2} \gamma_{xy}$: Half the change in angle between originally perpendicular lines in the x and y directions.
* $\varepsilon_{yz} = \varepsilon_{zy} = \varepsilon_{23} = \varepsilon_{32} = \frac{1}{2} \left( \frac{\partial u_y}{\partial z} + \frac{\partial u_z}{\partial y} \right) = \frac{1}{2} \gamma_{yz}$: Half the change in angle between originally perpendicular lines in the y and z directions.
* $\varepsilon_{zx} = \varepsilon_{xz} = \varepsilon_{31} = \varepsilon_{13} = \frac{1}{2} \left( \frac{\partial u_z}{\partial x} + \frac{\partial u_x}{\partial z} \right) = \frac{1}{2} \gamma_{zx}$: Half the change in angle between originally perpendicular lines in the x-direction.

where $u_x, u_y, u_z$ (or $u_1, u_2, u_3$) are the components of the displacement vector.

In matrix form, the deformation tensor can be written as:

$$
\boldsymbol{\varepsilon} = \begin{pmatrix}
\varepsilon_{xx} & \varepsilon_{xy} & \varepsilon_{xz} \\
\varepsilon_{yx} & \varepsilon_{yy} & \varepsilon_{yz} \\
\varepsilon_{zx} & \varepsilon_{zy} & \varepsilon_{zz}
\end{pmatrix} =
\begin{pmatrix}
\\
\\
\frac{1}{2} \left( \frac{\partial u_z}{\partial x} + \frac{\partial u_x}{\partial z} \right) & \frac{1}{2} \left( \frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z} \right) & \frac{\partial u_z}{\partial z}
\end{pmatrix}
$$

>warning
>
>This definition applies to **small deformations**.


## Hooke's law

For a linearly elastic material, Hooke's law applies in three dimensions:

$$\varepsilon_x = \frac{1}{E} [\sigma_x - \nu (\sigma_y + \sigma_z)]$$

$$\varepsilon_y = \frac{1}{E} [\sigma_y - \nu (\sigma_x + \sigma_z)]$$

$$\varepsilon_z = \frac{1}{E} [\sigma_z - \nu (\sigma_x + \sigma_y)]$$

$$\gamma_{xy} = \frac{\tau_{xy}}{G}$$

$$\gamma_{yz} = \frac{\tau_{yz}}{G}$$

$$\gamma_{zx} = \frac{\tau_{zx}}{G}$$

where
* $E$ is Young's modulus of elasticity,
* $\nu$ is Poisson's number and
* $G = \frac{E}{2(1+\nu)}$ is the shear modulus of elasticity.

### General linear elastic material

In a linearly elastic material, Hooke's law describes the relationship between the stress tensor $\boldsymbol{\sigma}$ and the strain tensor $\boldsymbol{\varepsilon}$. In the general three-dimensional state of stress and strain, this relationship can be expressed in tensor form as:

$$
\boldsymbol{\varepsilon} = \mathbf{S} : \boldsymbol{\sigma}
$$

where $\mathbf{S}$ is the **compliance tensor** (often also denoted as $\mathbf{C}^{-1}$, where $\mathbf{C}$ is the stiffness tensor). The colon ($:$) denotes a double contraction of tensors.

In matrix form, using Voigt's notation for the stress vector $\{\sigma\} = [\sigma_x, \sigma_y, \sigma_z, \tau_{xy}, \tau_{yz}, \tau_{zx}]^T$ and the strain vector $\{\varepsilon\} = [\varepsilon_x, \varepsilon_y, \varepsilon_z, \gamma_{xy}, \gamma_{yz}, \gamma_{zx}]^T$, Hooke's law is written as:

$$
\begin{pmatrix}
\varepsilon_x \\
\varepsilon_y \\
\varepsilon_z \\
\gamma_{xy} \\
\gamma_{yz} \\
\gamma_{zx}
\end{pmatrix} =
\begin{pmatrix}
S_{11} & S_{12} & S_{13} & S_{14} & S_{15} & S_{16} \\
S_{21} & S_{22} & S_{23} & S_{24} & S_{25} & S_{26} \\
S_{31} & S_{32} & S_{33} & S_{34} & S_{35} & S_{36} \\
S_{41} & S_{42} & S_{43} & S_{44} & S_{45} & S_{46} \\
S_{51} & S_{52} & S_{53} & S_{54} & S_{55} & S_{56} \\
S_{61} & S_{62} & S_{63} & S_{64} & S_{65} & S_{66}
\end{pmatrix}
\begin{pmatrix}
\sigma_x \\
\sigma_y \\
\sigma_z \\
\tau_{xy} \\
\tau_{yz} \\
\tau_{zx}
\end{pmatrix}
$$

The compliance matrix $[S]$ is symmetric, meaning $S_{ij} = S_{ji}$, and thus has at most 21 independent constants for a generally anisotropic material.

**For an isotropic material** (whose properties are the same in all directions), the compliance matrix is greatly simplified and is expressed using two independent material constants, Young's modulus of elasticity ($E$) and Poisson's ratio ($\nu$), or using Young's modulus ($E$) and shear modulus ($G = \frac{E}{2(1+\nu)}$), or using Lamé's constants ($\lambda$ and $\mu = G$). The compliance matrix for an isotropic material has the form:

$$
[S] =
\begin{pmatrix}
1/E & -\nu/E & -\nu/E & 0 & 0 & 0 \\
-\nu/E & 1/E & -\nu/E & 0 & 0 & 0 \\
-\nu/E & -\nu/E & 1/E & 0 & 0 & 0 \\
0 & 0 & 0 & 1/G & 0 & 0 \\
0 & 0 & 0 & 0 & 1/G & 0 \\
0 & 0 & 0 & 0 & 0 & 1/G
\end{pmatrix}
$$

where diagonal elements are related to normal strains induced by normal stresses and non-diagonal elements (off-zero) are related to transverse strains (Poisson's effect) and shear strains induced by shear stresses.

$$
[S] = \frac{1}{E}
\begin{bmatrix}
1 & -\nu & -\nu & 0 & 0 & 0 \\
-\nu & 1 & -\nu & 0 & 0 & 0 \\
-\nu & -\nu & 1 & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+\nu) & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+\nu) & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+\nu)
\end{bmatrix}
$$

## 11. Stiffness matrix

$$
[K] = \frac{E}{(1+\nu)(1-2\nu)}
\begin{bmatrix}
1-\nu & \nu & \nu & 0 & 0 & 0 \\
\nu & 1-\nu & \nu & 0 & 0 & 0 \\
\nu & \nu & 1-\nu & 0 & 0 & 0 \\
0 & 0 & 0 & \frac{1-2\nu}{2} & 0 & 0 \\
0 & 0 & 0 & 0 & \frac{1-2\nu}{2} & 0 \\
0 & 0 & 0 & 0 & 0 & \frac{1-2\nu}{2}
\end{bmatrix}
$$

## Main voltage

Consider a planar stress state (2D), where the stress tensor $\boldsymbol{\sigma}$ is reduced to a 2x2 matrix:

$$
\boldsymbol{\sigma} = \begin{pmatrix}
\sigma_x & \tau_{xy} \\
\tau_{yx} & \sigma_y
\end{pmatrix}
$$

Due to the moment equilibrium, $\tau_{xy} = \tau_{yx}$ holds, so the stress tensor is symmetric.

The principal stresses ($\sigma_1, \sigma_2$) are the eigenvalues of the stress tensor. They can be found by solving the characteristic equation:

$$
\det(\boldsymbol{\sigma} - \lambda \mathbf{I}) = 0
$$

where $\lambda$ represents the principal stress and $\mathbf{I}$ is a 2x2 identity matrix. By writing out the determinant:

$$
\det \begin{pmatrix}
\sigma_x - \lambda & \tau_{xy} \\
\tau_{xy} & \sigma_y - \lambda
\end{pmatrix} = (\sigma_x - \lambda)(\sigma_y - \lambda) - (\tau_{xy})^2 = 0
$$

Expanding this equation, we get the quadratic equation for $\lambda$:

$$
\lambda^2 - (\sigma_x + \sigma_y) \lambda + (\sigma_x \sigma_y - \tau_{xy}^2) = 0
$$

Solving this quadratic equation using the formula for roots gives the principal stresses $\sigma_1$ and $\sigma_2$:

$$
\lambda_{1,2} = \frac{(\sigma_x + \sigma_y) \pm \sqrt{(\sigma_x + \sigma_y)^2 - 4(\sigma_x \sigma_y - \tau_{xy}^2)}}{2}
$$

$$
\lambda_{1,2} = \frac{(\sigma_x + \sigma_y) \pm \sqrt{\sigma_x^2 + 2\sigma_x \sigma_y + \sigma_y^2 - 4\sigma_x \sigma_y + 4\tau_{xy}^2}}{2}
$$

$$
\lambda_{1,2} = \frac{(\sigma_x + \sigma_y) \pm \sqrt{\sigma_x^2 - 2\sigma_x \sigma_y + \sigma_y^2 + 4\tau_{xy}^2}}{2}
$$

$$
\lambda_{1,2} = \frac{(\sigma_x + \sigma_y) \pm \sqrt{(\sigma_x - \sigma_y)^2 + (2\tau_{xy})^2}}{2}
$$

Let us denote the principal stresses as $\sigma_1$ and $\sigma_2$:

$$
\sigma_1 = \frac{(\sigma_x + \sigma_y)}{2} + \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2}
$$

$$
\sigma_2 = \frac{(\sigma_x + \sigma_y)}{2} - \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2}
$$

Now we introduce the average normal stress $\sigma_m$ and the radius of the circle $R$:

$$
\sigma_m = \frac{\sigma_x + \sigma_y}{2}
$$

$$
R = \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2}
$$

Using these definitions, the principal stress can be written as:

$$
\sigma_1 = \sigma_m + R
$$

$$
\sigma_2 = \sigma_m - R
$$

Now let us focus on the stress ($\sigma_n, \tau_t$) acting on a plane rotated by an angle $\theta$ from the $x$ axis. The stress transformation for the plane stress state is given by the equations:

$$
\sigma_n = \frac{\sigma_x + \sigma_y}{2} + \frac{\sigma_x - \sigma_y}{2} \cos(2\theta) + \tau_{xy} \sin(2\theta)
$$

$$
\tau_t = -\frac{\sigma_x - \sigma_y}{2} \sin(2\theta) + \tau_{xy} \cos(2\theta)
$$

Our goal is to show that the points $(\sigma_n, \tau_t)$ lie on a circle in the plane $\sigma - \tau$. Let us rearrange the transformation equations:

$$
\sigma_n - \frac{\sigma_x + \sigma_y}{2} = \frac{\sigma_x - \sigma_y}{2} \cos(2\theta) + \tau_{xy} \sin(2\theta)
$$

$$
\tau_t = -\frac{\sigma_x - \sigma_y}{2} \sin(2\theta) + \tau_{xy} \cos(2\theta)
$$

Now we square both equations and add them:

$$
\left(\sigma_n - \frac{\sigma_x + \sigma_y}{2}\right)^2 + \tau_t^2 = \left( \frac{\sigma_x - \sigma_y}{2} \cos(2\theta) + \tau_{xy} \sin(2\theta) \right)^2 + \left( -\frac{\sigma_x - \sigma_y}{2} \sin(2\theta) + \tau_{xy} \cos(2\theta) \right)^2
$$

By expanding the right side:

$$
\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 \cos^2(2\theta) + \tau_{xy}^2 \sin^2(2\theta) + 2 \frac{\sigma_x - \sigma_y{\right) \sin^2(2\theta) + \tau_{xy}^2 \cos^2(2\theta) - 2 \frac{\sigma_x - \sigma_y}{2} \tau_{xy} \sin(2\theta) \cos(2\theta)
$$

The middle terms are subtracted, leaving:

$$
\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 (\cos^2(2\theta) + \sin^2(2\theta)) + \tau_{xy}^2 (\sin^2(2\theta) + \cos^2(2\theta))
$$

Using the trigonometric identity $\sin^2(\alpha) + \cos^2(\alpha) = 1$ we get:

$$
\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2
$$

The left side of the equation is:

$$
\left(\sigma_n - \frac{\sigma_x + \sigma_y}{2}\right)^2 + \tau_t^2
$$

By combining both sides:

$$
\left(\sigma_n - \frac{\sigma_x + \sigma_y}{2}\right)^2 + \tau_t^2 = \left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2
$$

Substituting the previously defined average normal stress $\sigma_m = \frac{\sigma_x + \sigma_y}{2}$ and the radius of the circle $R = \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2}$ we obtain the equation of the circle:

$$
(\sigma_n - \sigma_m)^2 + \tau_t^2 = R^2
$$

This equation represents **Mohr's circle** in the $\sigma - \tau$ plane, where:

* The center of the circle lies on the normal stress axis $\sigma$ at the point $(\sigma_m, 0) = \left(\frac{\sigma_x + \sigma_y}{2}, 0\right)$.
* The radius of a circle is $R = \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2}$.

The principal stresses $\sigma_1$ and $\sigma_2$ correspond to the intersections of the Mohr circle with the normal stress axis ($\tau_t = 0$), where $\sigma_n = \sigma_m \pm R$. The maximum shear stress $\tau_{max}$ corresponds to the upper and lower points of the circle and has a magnitude equal to the radius $R$:

$$
\tau_{max} = R = \sqrt{\left(\frac{\sigma_x - \sigma_y}{2}\right)^2 + \tau_{xy}^2} = \frac{\sigma_1 - \sigma_2}{2}
$$

We have thus derived the Mohr's circle equation as a geometric representation of the principal numbers (and generally transformed components) of the stress tensor in 2D.

## Principal stresses and strains in 3D

We can proceed similarly in three dimensions.

$$
\det(\boldsymbol{\sigma} - \lambda \mathbf{I}) = 0
$$

$$
\det(\boldsymbol{\varepsilon} - \lambda \mathbf{I}) = 0
$$

In this case, solving the determinant leads to a cubic equation that has three solutions: the three principal stresses $\sigma_1, \sigma_2, \sigma_3$ or $\epsilon_1, \epsilon_2, \epsilon_3$.

## Volumetric deformation is a trace of the deformation tensor

Consider an infinitesimal volume element in the shape of a cuboid with initial dimensions $dx_1, dx_2, dx_3$. Its initial volume is $dV_0 = dx_1 \, dx_2 \, dx_3$.

After deformation, the dimensions of the element change. For **small deformations**, we can approximate the change in edge length in the direction of the $x_i$ axis as $(1 + \varepsilon_{ii}) dx_i$, where $\varepsilon_{ii}$ is the normal component of the deformation tensor in the $x_i$ direction.

The deformed dimensions of the element are therefore approximately:
* $dx'_1 = (1 + \varepsilon_{11}) dx_1$
* $dx'_2 = (1 + \varepsilon_{22}) dx_2$
* $dx'_3 = (1 + \varepsilon_{33}) dx_3$

The volume of the deformed element $dV$ is given by the product of these deformed dimensions (we neglect higher-order terms for small deformations):

$$
dV = dx'_1 \, dx'_2 \, dx'_3 = (1 + \varepsilon_{11}) dx_1 \, (1 + \varepsilon_{22}) dx_2 \, (1 + \varepsilon_{33}) dx_3
$$

$$
dV = (1 + \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33} + \text{higher order terms}) \, dx_1 \, dx_2 \, dx_3
$$

For small deformations, we neglect higher order terms:

$$
dV \approx (1 + \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}) \, dV_0
$$

Volumetric deformation $\varepsilon_v$ is defined as the relative change in volume:

$$
\varepsilon_v = \frac{dV - dV_0}{dV_0} = \frac{(1 + \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}) dV_0 - dV_0}{dV_0}
$$

$$
\varepsilon_v = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}
$$

The trace of the deformation tensor $\text{Tr}(\boldsymbol{\varepsilon})$ is the sum of its diagonal elements:

$$
\text{Tr}(\boldsymbol{\varepsilon}) = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}
$$

Comparing both expressions, we get:

$$
\varepsilon_v = \text{Tr}(\boldsymbol{\varepsilon})
$$

We have thus proven that for small deformations, the volume deformation is equal to the trace of the deformation tensor.


## Volumetric stress component and bulk modulus of elasticity

> The bulk modulus of elasticity $K$ characterizes the resistance of a material to volume change under uniform omnidirectional loading (hydrostatic pressure). It is defined as the ratio between the volume component of stress ($\sigma_m$) and the volume strain ($\varepsilon_v$):

> $$
K = \frac{\sigma_m}{\varepsilon_v}
$$

For a linearly elastic isotropic material, Hooke's law applies:

$$
\varepsilon_{xx} = \frac{1}{E} [\sigma_{xx} - \nu (\sigma_{yy} + \sigma_{zz})] \\
\varepsilon_{yy} = \frac{1}{E} [\sigma_{yy} - \nu (\sigma_{xx} + \sigma_{zz})] \\
\varepsilon_{zz} = \frac{1}{E} [\sigma_{zz} - \nu (\sigma_{xx} + \sigma_{yy})]
$$

where $E$ is Young's modulus of elasticity and $\nu$ is Poisson's number.

The volume deformation $\varepsilon_v$ is the trace of the deformation tensor:

$$
\varepsilon_v = \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz}
$$

By adding the equations of Hooke's law for normal deformations, we get:

$$
\varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz} = \frac{1}{E} [(\sigma_{xx} + \sigma_{yy} + \sigma_{zz}) - 2\nu (\sigma_{xx} + \sigma_{yy} + \sigma_{zz})]
$$

$$
\varepsilon_v = \frac{1}{E} (1 - 2\nu) (\sigma_{xx} + \sigma_{yy} + \sigma_{zz})
$$

Volumetric stress component $\sigma_m$ (for hydrostatic loading $\sigma_{xx} = \sigma_{yy} = \sigma_{zz} = -p$, where $p$ is pressure, but in general):

$$
\sigma_m = \frac{1}{3} (\sigma_{xx} + \sigma_{yy} + \sigma_{zz})
$$

From the equation for volumetric deformation, we express the sum of normal stresses:

$$
\sigma_{xx} + \sigma_{yy} + \sigma_{zz} = E \frac{\varepsilon_v}{1 - 2\nu}
$$

We substitute this expression into the definition of the volume component of stress:

$$
\sigma_m = \frac{1}{3} \left( E \frac{\varepsilon_v}{1 - 2\nu} \right)
$$

Now we use the definition of the bulk modulus of elasticity $K = \frac{\sigma_m}{\varepsilon_v}$:

$$
K = \frac{\frac{1}{3} E \frac{\varepsilon_v}{1 - 2\nu}}{\varepsilon_v}
$$

After truncating $\varepsilon_v$ we obtain the resulting relation:

$$
K = \frac{E}{3(1 - 2\nu)}
$$


$$
\varepsilon_v = \text{Tr}(\boldsymbol{\varepsilon})
$$
