# Torsional Stress

> A shaft is a rotating machine component that is used to transmit torque from one location to another.

<div class='center-align'><figure><img src="https://engineeringlearn.com/wp-content/uploads/2021/09/Shafts-1024x539.jpg" width="500" ></figure></div>

When a circular cross-section is twisted, a torsional moment is exerted on the material, which induces shear stress in all the fibers of the cross-section. This phenomenon can be analyzed using the relations for torque and polar moment of inertia.

## Torque $ M_k $

<div class='center-align'><figure><img src="https://www.engineeringskills.com/images/posts/introduction-to-torsion/img6-1024x588.jpg" width="500" ></figure></div>

> Torque $ M_k $ is the moment that acts on a body about its axis, causing the body to rotate.

For a circular cross-section, this moment is defined as:

$$
M_k = \int\limits_A \rho \tau \, dA
$$

where:
- $ M_k $ is the torque,
- $ A $ is the cross-sectional area,
- $ \rho $ is the distance from the axis of rotation (radius),
- $ \tau $ is the shear stress at a given point in the cross-section.

## Torsional stress

We assume that the torsional stress is minimal at the shaft axis and increases linearly towards the edge:

$$\tau = c \rho $$

For the relationship between stress and torque, we must have

$$
M_k = \int\limits_A c \rho^2 \, dA
$$

from which we have

$$
c = \frac{M_k}{J_p}
$$
where $J_p$ is the polar moment of inertia of the cross section.

To calculate the shear stress $ \tau $ in the fibers of the cross section under the action of a torque, we use the relationship:

$$
\tau = \frac{M_k r}{J_p}
$$

where:
- $ \tau $ is the shear stress at a distance $ r $ from the axis of rotation,
- $ M_k $ is the torque,
- $ r $ is the distance from the axis of rotation (radius at a given point),
- $ J_p$ is the polar quadratic moment.

This relationship shows that the shear stress is directly proportional to the distance from the axis of rotation and the magnitude of the torque and inversely proportional to the polar moment of inertia.

### Maximum Stress

<div class='center-align'><figure><img src="https://www.engineeringskills.com/_next/image?url=%2Fimages%2Fposts%2Fintroduction-to-torsion%2Fimg5.jpg&w=384&q=75" width="300" ></figure></div>

The maximum shear stress occurs at the surface of the cross-section where $ r = R $ (the radius of the circle). The maximum shear stress is therefore:

$$
\tau_{\text{max}} = \frac{M_k \cdot R}{J_p}
$$

where:
- $ R $ is the radius of the circular cross-section.

## Polar moment of inertia $J_p$

> The polar moment of inertia $ J_p $ is the geometric moment that describes the resistance of the cross-section to torsion.

For a circular cross-section it is defined as:

$$
J_p = \int\limits_A \rho^2 \, dA
$$

For a circular cross-section of radius $ R $, the polar moment of inertia is calculated as:

$$
J_p = \frac{\pi R^4}{2}
$$

where:
- $ R $ is the radius of the circle.
