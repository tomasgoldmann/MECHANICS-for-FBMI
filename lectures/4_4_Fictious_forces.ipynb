# Fictious forces
Textbooks often state that "centrifugal force does not exist." When a satellite orbits the Earth, it is not held in balance by two equal and opposite forces, gravity pulling inward and centrifugal force pulling outward. The satellite tends to move in a straight line, and there is only one force, gravity, that curves its path.

> **note**
>
>The word centrifugal comes from the Latin words centrum, "center," and fugere, "to flee," so the word means "fleeing from the center." Centrifugal force was studied by physicists as early as 1629, and the term itself was used by Sir Isaac Newton in its Latin form vis centrifuga in 1687.

On the other hand, we don't feel centripetal force much, but when we drive a car around a bend too fast and feel ourselves being thrown away from the center of curvature of our path, the "centrifugal force" acting on us seems very real. The whole problem that arose is in the description of our reference system.

## Inertial reference frame

An inertial reference frame is a reference frame in which the law of inertia applies. Any reference frame, if it is at rest or in uniform rectilinear motion with respect to a given inertial frame, is also inertial.
Systems in which the law of inertia does not apply are called non-inertial. They move in a straight line, accelerated, decelerated, or curved motion.

## D'Alembert's Principle

Let's look at a simpler example that does not involve rotation. A car accelerates with an acceleration of $\vec{a}$ to the right.

The suspended plumb line in the car is not in a vertical position due to acceleration. Some would say that there are only two forces acting on the plumb line – its weight and the tension in the string – and as a result, the plumb line accelerates to the right. Using the relation $\vec{F}=m\vec{a}$, it is easy to find the tension in the string and the angle that the string makes with the vertical.

![](https://phys.libretexts.org/@api/deki/files/6461/figure_IV.20.png?revision=1&size=bestfit&width=486&height=240)

However, the passenger in the car sees things differently. For the passenger in the car, nothing is accelerating. The plumb line is only in static equilibrium under the action of three forces, one of which is the force $ma$ to the left.

![](https://phys.libretexts.org/@api/deki/files/6462/figure_IV.21..png?revision=1&size=bestfit&width=382&height=227)

It seems that in the frame of reference outside the car there are only two forces, but in the accelerating frame of reference inside the car the system can be perfectly described by postulating the existence of the force $ma$ pulling to the left. This is a principle of classical mechanics, known as **d'Alembert's principle**, where if we describe a system in an accelerating frame of reference, we can replace the acceleration by a force in the opposite direction. D'Alembert's principle, an alternative formulation of Newton's second law of motion, was formulated by the 18th-century French polymath Jean Le Rond d'Alembert. Essentially, the principle reduces a problem in dynamics to a problem in statics. The second law states that the force $F$ acting on a body is equal to the product of the mass $m$ and the acceleration a of the body, or $F = ma$; in d'Alembert's form, the force $F$ plus the negative of the mass $m$ times the acceleration a of the body equals zero: $F − ma = 0$. In other words, the body is in equilibrium under the action of the real force F and the fictitious force −ma. The fictitious force is also called the inertial force and the inverse effective force.

> D'Alembert's principle states that the sum of all forces acting on a system, including inertial forces, is zero. Inertial forces are fictitious forces that appear in non-inertial (accelerating or rotating) frames of reference.

Mathematically, D'Alembert's principle can be expressed as:

$$\sum (\vec{F}_i - m_i \vec{a}_i) \cdot \delta \vec{r}_i = 0$$

where:

* $\vec{F}_i$ is the resultant external force acting on the ith particle.
* $m_i$ is the mass of the ith particle.
* $\vec{a}_i$ is the acceleration of the ith particle.
* $\delta \vec{r}_i$ is the virtual displacement of the ith particle.

Imagine a block on an inclined plane. According to Newton's second law, the block accelerates down the plane due to the force of gravity.

#### Inertial reference frame (Newton's law)

* Forces acting on the block:
* Gravitational force ($\vec{F}_g$)
* Normal force ($\vec{N}$)
* Frictional force ($\vec{F}_f$)
* The resultant force causes the block to accelerate ($\vec{a}$).

#### Non-inertial reference frame (D'Alembert's principle)

* Forces acting on the block:
* Gravitational force ($\vec{F}_g$)
* Normal force ($\vec{N}$)
* Frictional force ($\vec{F}_f$)
* Inertial force ($\vec{F}_i = -m\vec{a}$)
* According to D'Alembert's principle, the sum of these forces is zero, which means that the block is in equilibrium.

D'Alembert's principle is important because:
* It allows us to solve complex dynamic problems using the methods of statics.
* It is the basis of Lagrangian mechanics, which is a powerful tool for analyzing complex mechanical systems.
* It helps us understand the concept of inertial forces and their effect on motion.
A similar situation exists with centrifugal force. If you consider a satellite in orbit around the Earth, some would say that there is only one force acting on the satellite, the gravitational force toward the Earth. A satellite in a circular orbit accelerates toward the center of the circle, and everything is as expected - $F=ma$. The acceleration is centripetal acceleration. An astronaut aboard the satellite may have a different view. He is at a constant distance from the Earth, not accelerating; he is in static equilibrium and feels no resultant force on himself - he feels weightless. If he has been taught that the Earth exerts a gravitational force, then this must be balanced by a force directed away from the Earth. This force, which will manifest itself when described in a rotating frame of reference, is the centrifugal force.

## Rotating coordinate systems

* $\Sigma = Oxyz$ is an inertial frame of reference (not accelerating or rotating).
* $\Sigma' = Ox'y'z'$ is a frame of reference that rotates about the $z$ axis with angular velocity $\vec{\omega} = \omega \vec{k}$.

Three questions:

1. If point P is fixed with respect to $\Sigma'$, what is its velocity $v$ with respect to $\Sigma$?
2. If point P moves with velocity $v'$ with respect to $\Sigma'$, what is its velocity $v$ with respect to $\Sigma$?
3. If point P has acceleration $a'$ with respect to $\Sigma'$, what is its acceleration $a$ with respect to $\Sigma$?

Consider a simple 2D problem where the $z$ axes are the same and the system rotates with velocity $\vec{\omega}$

1. The answer to the first question is:

$$\vec{v} = \vec{\omega} \times \vec{r}$$

2. If P is moving with velocity $v'$ relative to $\Sigma'$, then its velocity $v$ relative to $\Sigma$ is:

$$\vec{v} = \vec{v}' + \vec{\omega} \times \vec{r}$$

or

$$\dot{\vec{r}} = \dot{\vec{r}}' + \vec{\omega} \times \vec{r}$$

This shows that:

$$\left(\frac{d}{dt}\right)_\Sigma = \left(\frac{d}{dt}\right)_{\Sigma'} + \vec{\omega} \times$$

3. If P is accelerating relative to $\Sigma'$, we can use the operator k equation:

 $${\vec{a}} = (\frac{d}{dt})_{ \Sigma^{\prime}} ({\vec v}^{\prime} + \vec \omega \times {\vec r}) + \vec \omega \times ({\vec v}^{\prime} + \vec \omega \times {\vec r}) $$

$$= {\vec a}^{\prime} +\vec \omega \times \vec v^{\prime} + \vec \omega \times \vec v^{\prime} + \vec \omega \times ( \vec \omega \times {\vec r} ).$$

$$\vec{a} = {\vec a}^{\prime} + \vec \omega \times( \vec \omega \times {\vec r}) + 2 \vec \omega\times \vec{v^{\prime}}$$

If P is neither accelerating nor moving relative to $\Sigma'$ , then:

$$a = \omega \times (\omega \times r)$$

Which is the centripetal acceleration as we can easily prove by ${\bf A} \times({ \bf B} \times { \bf C} )=({ \bf A} \cdot { \bf C)B-(A\cdot B)C}$.

$$\vec a = \vec \omega \times (\vec \omega \times \vec r) = (\vec \omega \cdot \vec r)\vec \omega - \omega^2 \vec r$$

If a particle moves with a velocity $\vec v'$ relative to $\Sigma'$, it has an additional acceleration $2\vec \omega \times \vec v'$ relative to $\Sigma$, which is the **Coriolis acceleration**.

The inverse equation is:

$$\vec a' = \vec a + \vec \omega \times (\vec r \times \vec \omega) + 2\vec v' \times \vec \omega$$

$$\vec F' = \vec F + m\vec \omega \times (\vec r \times \vec \omega) + 2m \vec v' \times \vec \omega$$

Coriolis Force:

$$F = 2m\vec v' \times \vec \omega$$

## Coriolis Force - Thought Experiment

Now imagine a thought experiment (*Gedanken Prüfung*). Go to the equator and build a tall tower. Drop a stone from the top of the tower. Think about the direction of the vector $2m\vec{v}' \times \vec{\omega}$. Really think about it. Or think about the law of conservation of angular momentum. The stone falls closer to the axis of rotation of the Earth. It must conserve angular momentum. It falls east of the tower (not west!).

![](https://phys.libretexts.org/@api/deki/files/6465/Figure_IV.24.png?revision=1&size=bestfit&width=578&height=309)

Two forces act on the stone: its weight $mg$ and the Coriolis force. The Earth's rotation vector $\vec{\omega}$ points north. The Coriolis force is perpendicular to the stone's velocity. If we decompose the stone's velocity into a vertical component $\dot{y}$ pointing down and a horizontal component $\dot{z}$ pointing east, the corresponding components of the Coriolis force will be $2m\omega\dot{y}$ pointing east and $2m\omega\dot{x}$ pointing up. However, I assume that $\dot{z} \ll \dot{y}$ and the only significant Coriolis force is the eastward component $2m\omega\dot{y}$ that I have drawn. Another way to express this approximation is to say that the upward component of the Coriolis force is negligible compared to the weight $mg$ of the stone.

After falling for $t$, the $y$ coordinate of the stone is determined in the usual way:

$$\ddot{y} = g, \quad \dot{y} = gt, \quad y = \frac{1}{2}gt^2$$

And the $x$ coordinate is determined from:

$$\ddot{x} = 2\omega\dot{y} = 2\omega gt, \quad \dot{x} = \omega gt^2, \quad x = \frac{1}{3}\omega gt^3$$

This way you can find out how far east the stone fell after two seconds, or how far east it fell if the height of the tower is 100 meters. The equation of the trajectory would be the $t$-eliminant:

$$x^2 = \frac{8\omega^2}{9g}y^3$$
For the Earth, $\omega = 7.292 \times 10^{-5} \, \text{rad s}^{-1}$ and at the equator $g = 9.780 \, \text{m s}^{-2}$, so:

$$\frac{8\omega^2}{9g} = 4.788 \times 10^{-10} \, \text{m}^{-1}$$
