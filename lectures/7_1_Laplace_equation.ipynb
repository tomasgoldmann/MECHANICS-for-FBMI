# Stress in thin-walled vessels

Laplace's equation for thin-walled vessels describes the relationship between the internal pressure $p$, the radius of curvature of the vessel, and the stress in the vessel wall. This relationship is important in biomechanics (e.g., for modeling blood vessels), mechanical engineering, and materials engineering.

The initial division of tasks is done according to the ratio of wall thickness to the total size of the vessel. If the smallest internal radius of curvature of the vessel $R_{min}$ and the wall thickness $t$ are indicated, we can indicate
* $\frac{R_{min}}{t} < 10$ - we are talking about thin-walled vessels
* $\frac{R_{min}}{t} > 10$ - we are talking about thick-walled vessels

For thin-walled vessels, we do not need the inner and outer radii, but the average radius is sufficient for the calculation. Due to the small wall thickness, the stress in thin-walled vessels is considered to be planar (biaxial), even though the pressure p acts on the inner surface. Since the occurrence of shear stresses in the wall is not considered, the two arising stresses will be the main stresses of the considered stress, which is also called membrane.

We will consider only **thin-walled vessels**. The issue of stress in thick-walled vessels goes beyond the scope of our course. Let us assume a vessel with two main radii of curvature $R_1$ and $R_2$ (they lie in two mutually perpendicular planes that correspond to the directions of the main stress, these are the so-called main curvatures), a wall thickness $t$ and an internal pressure $p$.

>**note**
>
>For thin-walled vessels, we do not need to consider the inner and outer radii, because $t\ll R_1,E_2$. Due to the small wall thickness, the stress in thin-walled vessels is considered to be planar (biaxial), even if the pressure $p$ acts on the inner surface. Since the occurrence of shear stresses in the wall is not considered, the two arising stresses will be directly the principal stresses of the considered stress, which is also called **membrane stress**.


The relationship between the stress in the wall and the shape of the vessel is referred to as the **Young-Laplace equation**. This equation was derived approximately simultaneously by **Thomas Young (1804)** and **Simon Pierre de Laplace (1805)**. A brief derivation of this equation is given below.

## Derivation based on force equlibrium

![](https://www.researchgate.net/profile/Svein-Skjaeveland/publication/351774715/figure/fig3/AS:1026185635037185@1621673129539/Displaced-surface-between-two-phases_Q320.jpg)

Assume that in the direction determined by curvature $R_1$ there is a stress of $\sigma_1$ and in the direction determined by curvature $R_2$ there is a stress of $\sigma_2$. From the force balance in direction 1 we get:

$$F_1 = 2 \sigma_1 R_2 \mathrm{d}\theta_2 t \sin \frac{\theta_1}{2} = \sigma_1 R_2 \mathrm{d}\theta_2 t \theta_1$$

and from the force balance in direction 2 we get

$$F_2 = 2 \sigma_2 R_1 \mathrm{d}\theta_1 t \sin \frac{\theta_2}{2} = \sigma_2 R_1 \mathrm{d}\theta_1 t \theta_2$$

where we used the fact that for a small membrane element $\theta \rightarrow 0$ $\sin \theta \approx \theta$ holds.

The equilibrium of forces in the direction normal to the surface of the element can be expressed as

$$F_1 + F_2 = p A $$

and for a small element, it is valid

$$ \sigma_1 R_2 \mathrm{d}\theta_2 t \theta_1 + \sigma_2 R_1 \mathrm{d}\theta_1 t \theta_2 = p R_1 \mathrm{d}\theta_1 R_2 \mathrm{d}\theta_2$$

From which it follows that the stress does not depend on the size of the element but can be described by the relation

$$\frac{\sigma_1}{R_1} + \frac{\sigma_2}{R_2} = \frac{p}{t} $$
where:
- $\sigma_{1,2}$ are the principal stresses in the vessel wall,
- $p$ is the internal pressure,
- $R_{1,2}$ are the principal radii of curvature of the wall container,
- $t$ is the thickness of the container wall,
-
Laplace's equation says that:
* The greater the pressure acting inside the container, the greater the stress in the wall.
* The larger the radius of the container, the greater the stress the wall must withstand.
* The thicker (thicker) the container wall, the less stress it generates.

:::{caution}
Principal stresses are normal stresses acting on the main planes of the material, where there are no shear stresses. We denote them as $\sigma_1$, $\sigma_2$ and $\sigma_3$ and rank them by magnitude

$$\sigma_1 \geq \sigma_2 \geq \sigma_3$$
For a thin-walled container, $\sigma_3 \approx 0$, because $t \ll R$. The remaining two stresses are tensile. Their value is determined by the radius.
:::

## Derivation based on strain energy

![]( https://i0.wp.com/d2ottx2pd20o0z.cloudfront.net/wp-content/uploads/2023/06/07142305/Laplace-1-300x288-1.webp?resize=300%2C288&ssl=1)

Consider a small element of a curved surface with Cartesian dimensions $x$ and $y$. On the normal $n$, two arcs can be constructed, having lengths $x$ and $y$ and radii $R_1$ and $R_2$.

It is valid
$$x = \theta_1 R_1$$
and
$$y = \theta_2 R_2$$

When the radii of curvature are increased by a small value $dR$, the work done is $dW$:

$$
dW = \sigma_1 t y dx + \sigma_2 t x dy
$$
where
$$dA = x dy + y dx $$
where:
- $dA$ is the increase in surface area.

This work is done by the pressure $p$. According to thermodynamics, it can also be expressed as:

$$
dW = p \, dV
$$

where $dV$ is the elementary change in the container.

$$ dV = x y dR$$

From these relations, the pressure $p$ can be expressed:

$$
p = \frac{\sigma_1 t y dx + \sigma_2 t x dy }{x y dR}
$$

$$
p = \sigma_1\frac{dx}{x dR}+ \sigma_2\frac{dy}{y dR}
$$
From geometry, we have

$$
\frac{dx}{dR} = \theta_1 = \frac{x}{R_1}
$$

and therefore

$$
\frac{p}{t} = \frac{\sigma_1}{R_1} + \frac{\sigma_2}{R_2}
$$

## Spherical vessel

For a thin-walled spherical vessel with radius $R$ and thickness $t$,

$$R_1 = R_2 = R $$

After modification, we get **Laplace's equation for a spherical vessel**:

$$
\sigma = \frac{p R}{2t}
$$

## Closed cylindrical vessel

Let us now consider a thin-walled cylindrical tube with radius $R$ and thickness $t$. The wall stress can be analyzed in two directions:

### 1. Longitudinal stress ($\sigma_z$)

The force balance in the direction of the vessel axis shows that the longitudinal stress must balance the force acting on the circular section:

$$
p \pi R^2 = \sigma_z 2\pi R t
$$

After adjustment:

$$
\sigma_z = \frac{p R}{2t}
$$

### 2. The circumferential (tangential) stress $\sigma_{\theta}$

can be directly determined from Laplace's equation, since $R_1=R, R_2=0$.

$$
\frac{p}{t} = \frac{\sigma_\theta}{R}
$$

$$
\sigma_\theta = \frac{p R}{t}
$$

## Applications of Laplace's equation

### Biomechanics of blood vessels

The following relationship applies to blood vessels:

$$
\sigma = \frac{p R}{t}
$$

This relationship explains:
- **Larger arteries must have thicker walls**, because for the same blood pressure, their radius is larger, and therefore the tension in the wall increases.
- **Aneurysm (bulging of a vessel)**: If a vessel expands locally (increases $R$), the tension in the wall increases, which leads to further expansion and possible rupture of the vessel.

### Lungs and alveoli

The alveoli in the lungs are almost spherical and their surface is covered with a thin film of liquid. The surface tension of this liquid creates a pressure that is described by Laplace's equation:

$$
p = \frac{2 \gamma}{R}
$$

where $\gamma$ is the surface tension. This means:
- **Smaller alveoli require more pressure to inflate**, which could cause them to collapse.
- **Surfactant (a biological substance in the lungs) reduces surface tension**, helping to maintain a balance between the different sized alveoli.

### Pressure vessels
- Cylindrical pressure vessels (e.g. gas cylinders) have more stress in the circumferential direction than in the longitudinal direction, which explains their design.
- Spherical pressure vessels (e.g. balloons) have a more even stress distribution, which is advantageous at high internal pressures.
