# Construction of Mohr's Circle

## **1. Initial Values**
First, the principal stress components at a given point in the material are determined:
- Normal Stresses $ \sigma_x $, $ \sigma_y $
- Shear Stress $ \tau_{xy} $

![](https://blog.prepineer.com/wp-content/uploads/2017/10/01-Mohrs-Cricle-Principal-Stresses.jpg)

## **2. Determining Points on Mohr's Circle**
- Point $ A(\sigma_x, \tau_{xy}) $ corresponds to the stress on the plane with the normal in the direction of the $ x $ axis.
- Point $ B(\sigma_y, -\tau_{xy}) $ corresponds to the stress on the plane with the normal in the direction of the $ y $ axis.

## **3. Circle construction**
1. Connect point $ A $ and point $ B $ with a line segment.
2. On the $ \sigma $ axis, mark the point $ C $ as the intersection of the line segment AB with the $\sigma$ axis.

## **4. Determining the principal stresses**
The principal stresses $ \sigma_1, \sigma_2 $ correspond to the intersections of the circle with the $ \sigma $ axis:

$$
\sigma_{1,2} = \frac{\sigma_x + \sigma_y}{2} \pm R
$$

## **7. Maximum Shear Stress**
The maximum shear stress corresponds to the top and bottom points of Mohr's circle:

$$
\tau_{\max} = R
$$

## **8. Determination of the principal and shear planes**
The angle of the principal stress planes with respect to the original coordinate system is given by:

$$
\tan 2\theta_p = \frac{2\tau_{xy}}{\sigma_x - \sigma_y}
$$

The angle between the $ x $ axis and the plane of maximum shear stress is:

$$
\theta_s = \theta_p + 45^\circ
$$

## Numerical determination of the center of the circle**
The center of the Mohr circle $ S $ has the coordinates:

$$
S = \left( \frac{\sigma_x + \sigma_y}{2}, 0 \right)
$$

## Numerical determination of the radius of the circle**
The radius $ R $ of the circle is given by:

$$
R = \sqrt{\left( \frac{\sigma_x - \sigma_y}{2} \right)^2 + \tau_{xy}^2}
$$

## Mohr's Circle Equation**
Mohr's circle is described by the equation:

$$
\left( \sigma - \frac{\sigma_x + \sigma_y}{2} \right)^2 + \tau^2 = R^2
$$
