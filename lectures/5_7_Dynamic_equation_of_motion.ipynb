# Dynamic equations of motion in a plane and rotational motion

Motions of bodies in a plane and rotational motions can be analyzed using **Newton's equations** and **d'Alembert's equations**, for both translational and rotational motion. These approaches allow the description of the forces and moments that determine the motion of a body, and are fundamental tools for the mechanics of rigid bodies.

## Newton's equations of motion in a plane

For translational motion in a plane, the following applies:

### Translational motion in a plane:

Newton's second law for motion in a plane is expressed as:

$$
\vec{F} = m \vec{a}
$$

Where:
- $ \vec{F} $ is the vector of all forces acting on the body,
- $ m $ is the mass of the body,
- $ \vec{a} $ is the acceleration of the body.

The body moves according to the components in the plane, i.e. in the $ x $ and $ y $ axes:

$$
F_x = m \frac{dv_x}{dt}, \quad F_y = m \frac{dv_y}{dt}
$$

### Rotational motion in the plane:

For rotational motion around a fixed axis in the plane, the analogy to translational motion applies, but instead of forces, a **moment of force** (torque) is applied, which causes rotation. Newton's rotational equation is:

$$
\vec{M} = I \vec{\alpha}
$$

Where:
- $ \vec{M} $ is the moment of force (torque) acting on the body,
- $ I $ is the moment of inertia of the body,
- $ \vec{\alpha} $ is the angular acceleration of the body.

## d'Alembert's equations of motion in a plane

d'Alembert's equations formulate motion in a plane and include both translational and rotational motion using inertial forces and moments. This approach is useful for complex systems where virtual work and complex interactions between forces and motion need to be included.

### Translational motion in a plane:

For translational motion, d'Alembert's equation is of the form:

$$
\vec{F} - m \vec{a} = 0
$$

where $ \vec{F} $ is the vector of forces acting on the body and $ m \vec{a} $ is the inertial force, which is negative because it is the force needed to accelerate the body. In a plane it will be:

$$
F_x - m \frac{dv_x}{dt} = 0, \quad F_y - m \frac{dv_y}{dt} = 0
$$

### Rotational motion in a plane:

For rotational motion we will use an analogous approach, where the d'Alembert equation is expressed for moments of force and moments of inertia:

$$
\vec{M} - I \vec{\alpha} = 0
$$
