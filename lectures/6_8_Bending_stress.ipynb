# Bending Stress

## Beam

> In structural engineering, a beam is a structural element that is designed to transmit bending loads perpendicular to the axis. This load can include both static loads (e.g., the weight of the structure) and dynamic loads (e.g., wind loads, movement).

The stress arising in a material during bending is caused by the bending moment that acts on the cross-section of the material. To calculate the bending stress, we use the bending moment $M_o$ and the squared moment of the section $J_y$, which characterizes the stiffness of the section against bending.

Let's imagine that we have an element that is bent by the bending moment $ M_o $. This moment induces stress in all fibers of the cross-section. The stress at a point in a cross-section depends on its distance from the neutral axis, which is the axis around which bending occurs.

## Neutral axis

> The neutral axis is an imaginary line or plane in a material that is not subjected to any normal stresses during bending.

There is no deformation in the sense of compression or extension at the neutral axisâ€”it is a location where the stress is zero. During bending, the parts of the material above the neutral axis are compressed (experiencing compressive stresses), while the parts below the neutral axis are stretched (experiencing tensile stresses).

### Neutral axis position

The position of the neutral axis is affected by the geometry of the cross-section and the material properties. For simple calculations, it is usually assumed that the neutral axis is parallel to the axis of symmetry of the cross-section (for example, the axis of the center of the cross-section for rectangular, circular, and other symmetrical cross-sections). For different types of cross-sections, the position of the neutral axis is determined by different factors.

1. **For symmetrical cross-sections (e.g. rectangular or circular cross-section)**
* The neutral axis is usually located at the center of the cross-section. For a rectangular cross-section, for example, the neutral axis passes through the center of the height $h$, i.e. at the point $y = \frac{h}{2}$.

2. **For non-symmetrical cross-sections**
* For non-symmetrical cross-sections, the neutral axis is also determined by the geometric and material properties of the cross-section. The neutral axis is often offset from the axis of symmetry, which means that different calculations are created that take this offset into account.

When a material is bent, the stress must transmit the bending moment and does not transmit the force. The magnitude of the moment of force from tensile and compressive stresses must be the same.

$$
\int F = \int_A \sigma(y) dA = 0
$$

which is equivalent to defining the centroid of the cross-section when the axis $y=0$ passes through the centroid. This implies that the neutral axis of a homogeneous material passes through the centroid of the cross-section.

## Bending strain and stress

During bending, the fibers above the neutral axis are compressed, while the fibers below the neutral axis are stretched. The bending strain in the material is directly proportional to the distance $ y $ from the neutral axis. The stress created in the material is given by the relationship:

Similar to torsion, we can assume

$$\sigma = c y $$
where $y$ is the distance from the neutral axis

>**caution**
>
>A negative distance indicates a negative stress (pressure) in the case of a positive bending moment.


$$
M_o = \int M = \int_A y \sigma(y) dA = \int_A y \sigma(y) dA = \int_A c y^2 dA = c J_p
$$

$$
\sigma = \frac{M_o \cdot y}{J_y}
$$

where:
- $ \sigma $ is the stress at a point on the cross-section,
- $ M_o $ is the bending moment acting on the element,
- $ y $ is the distance of the given point from the neutral axis,
- $ J_y $ is the quadratic moment of the cross-section about the $ y $ axis.

## Quadratic moment of section $ J_y $

> The quadratic moment of section $ J_y $ } (or Area moment of inertia $ I_xx) is a geometric property, it determines the resistance to bending.

The quadratic moment of section $ J_y $ is defined as the integral of the distance of each small element of area $ dA $ from the $ y $ axis squared:

$$
J_y = \int_A y^2 \, dA
$$

There are specific expressions for $ J_y $ for different cross-sectional geometries, e.g. for a rectangular cross-section of width $ b $ and height $ h $ is:

$$
J_y = \frac{b h^3}{12}
$$

For a circular cross-section of radius $ R $ is:

$$
J_y = \frac{\pi R^4}{4}
$$

## Maximum bending stress

The maximum stress occurs at the edge of the cross-section, where the distance from the neutral axis is $ y_{\text{max}} = R $ for a circular cross-section or $ y_{\text{max}} = \frac{h}{2} $ for a rectangular cross-section. The maximum stress is therefore:

$$
\sigma_{\text{max}} = \frac{M_o \cdot y_\text{max}}{J_y}
$$
