## Example: Force on the legs of a person standing on a railroad car as it passes through a curve

According to https://www.lehman.edu/faculty/anchordoqui/chapter21.pdf

**Problem:**

A person of mass $m$ stands on a railroad car that is passing through a straight curve of radius $R$ at speed $v$. Her center of gravity is at a height $L$ above the car, midway between her legs, which are $d$ apart. The person is facing in the direction of motion. What is the magnitude of the normal force acting on each leg?

**Solution:**

We begin by choosing a coordinate system and sketching the release. We decompose the contact force on the inner leg (closer to the center of motion) into a tangential component corresponding to static friction $F_{t1}$ and perpendicular to it into a component of the normal force $N_1$. Similarly, we decompose the contact force between the outer foot (farther from the center of motion) and the ground into a tangential component corresponding to static friction $F_{t2}$ and perpendicular to it into a normal force component $N_2$.

We use two dynamic equations of motion: for translational motion and for rotational motion about the center of gravity, noting that in this special case $\alpha = 0$ because the object does not rotate about the center of gravity. We assume that all forces act at the center of gravity.

#### Equation for translational motion:

- **Radial component:**

$$
- F_{t1} - F_{t2} + m \frac{v^2}{R} = 0
$$

- **Vertical component:**

$$
N_1 + N_2 - mg = 0
$$

#### Equation for rotational motion:

The rotational equation of motion is:

$$
\sum \vec{M_{cm}} = 0
$$

where we have chosen to determine the moments of force relative to the center of mass.

To begin, we calculate the moments of force about the center of mass, noting that the gravitational force does not contribute to the moment because it acts at the center of mass. We draw a moment of force diagram showing the point of application of the forces, the point about which we calculate the moment of force, and the vector $\vec{r}_{cm,1}$ from this point to the point of application of the forces.

##### Moment of force on the inside leg:

$$
\vec{M}_{cm,1} = \vec{r}_{cm,1} \times (\vec{F}_{t1} + \vec{N_1}) = (-\frac{d}{2} N_1 - L \vec{F}_{t1}) \vec{k}
$$

##### Moment of force on the outside leg:

$$
\vec{M}_{cm,2} = \vec{r}_{cm,2} \times (\vec{\vec{F}_{t2}} + \vec{N_2}) = (\frac{d}{2} N_2 + L \vec{F}_{t2}) \vec{k}
$$

The sum of the moments of force about the center of gravity must be zero:

$$
{M}_{cm_{\text{ext}}} = {M}_{cm,1} + {M}_{cm,2} = \frac{d}{2}(-N_1 + N_2) + L (\vec{F}_{t2} + \vec{F}_{t1}) \hat{\theta} = 0
$$

This leads to the condition:

$$
-\frac{d}{2} (N_1 - N_2) + L m \frac{v^2}{R} = 0
$$

After rewriting:

$$
N_2 - N_1 = \frac{2Lmv^2}{Rd}
$$

Next, we rewrite the vertical component:

$$
N_2 + N_1 = mg
$$

Adding the equations and dividing by two, we get:

$$
N_2 = \frac{1}{2} mg + \frac{2Lmv^2}{Rd}
$$

For $N_1$, we get:

$$
N_1 = \frac{1}{2} mg - \frac{2Lmv^2}{Rd}
$$

Result: we see that the normal force acting on the outside leg is greater than the normal force acting on the inside leg. We expect this result because as the speed $v$ increases we find that at the maximum speed $v_{\text{max}}$ the normal force on the inside leg goes to zero and we start to rotate. This maximum speed can be found by setting $N_1 = 0$ v which leads to:

$$
v_{\text{max}} = \sqrt{\frac{gRd}{4L}}
$$
In this problem we also did not assume a maximum value of the static shear force.
