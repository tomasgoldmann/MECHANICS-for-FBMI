## Dimensioning beams for combined bending - compression

When designing beams that are subjected to a combination of bending and compression, it is necessary to take into account the interaction between these two effects, which affect the stresses in the beam. This problem is common for beams that are simultaneously loaded with axial compression and bending moments. The calculation is complex, because both effects can have a synergistic effect on the material and therefore on the strength and stability of the beam.

### Step 1: Determining bending moments and pressure

For a combination of bending and compression, we have two load components:

1. **Bending moment**: $M$ (in the plane where the bending occurs)
2. **Axial force**: $P$

This pressure acts on the beam in the direction of its axis and generates normal stresses. The bending moment causes bending stresses due to the curvature of the beam.

### Step 2: Determine the stress in the beam

The total stress at any point in the beam is the combination of the compressive and bending stresses. The bending stress is calculated as:

$$
\sigma_{bend} = \frac{M y}{I}
$$

where:
- $\sigma_{bend}$ is the bending stress.
- $M$ is the bending moment.
- $y$ is the distance of the point from the neutral axis.
- $I$ is the moment of inertia of the beam cross-section.

The compressive stress is uniformly distributed over the beam cross-section and is given by:

$$
\sigma_{axial} = \frac{P}{A}
$$

where:
- $\sigma_{axial}$ is the compressive stress.
- $P$ is the axial pressure.
- $A$ is the cross-sectional area of ​​the beam.

### Step 3: Stress Combination

The total stress at a point in the beam is calculated as the sum of the stresses induced by bending and compression. The total stress at a distance $y$ from the neutral axis is:

$$
\sigma_{total} = \sigma_{axial} + \sigma_{bend} = \frac{P}{A} + \frac{M y}{I}
$$

If $\sigma_{total}$ is positive, it means that the material is in a state of tensile stress, if it is negative, the material is in compression.

### Step 4: Checking the stress limit

To verify the correctness of the design, it is necessary to compare the resulting stress with the material's ultimate strength. Usually, the ultimate stress $\sigma_{max}$ is used, which can be the yield strength or the material's ultimate strength.

$$
\sigma_{total} \leq \sigma_{max}
$$

If this stress is less than the allowable stress, the design is safe. If the stress is greater, it is necessary to adjust the dimensions of the beam or change the material.
