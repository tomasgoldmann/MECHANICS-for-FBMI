# Steiner's theorem

## Definition of Steiner's theorem

Steiner's theorem (also known as the axis displacement theorem) allows us to calculate the moment of inertia of a body about any axis, if the moment of inertia about the axis passing through the center of gravity is known.

Mathematically, it is expressed as:

$$
I = I_C + M d^2
$$

where:
- $ I $ is the moment of inertia about the new axis,
- $ I_C $ is the moment of inertia about the axis passing through the center of gravity,
- $ M $ is the total mass of the body,
- $ d $ is the distance between the center of gravity axis and the new axis.

## Derivation of Steiner's theorem

Let us assume a rigid object composed of small mass elements $ dm $, whose moment of inertia we want to determine about an axis displaced by a distance $ d $ from the center of gravity.

Each mass element $ dm $ has coordinates $ x_C, y_C $ in the centroid system and $ x, y $ in the new system. The relationship between them is:

$$
x = x_C + d
$$

The moment of inertia about the new axis is:

$$
I = \int x^2 \, dm
$$

Substitute $ x = x_C + d $:

$$
I = \int (x_C + d)^2 \, dm
$$

Express the square:

$$
I = \int (x_C^2 + 2 d x_C + d^2) \, dm
$$

Divide the integral:

$$
I = \int x_C^2 \, dm + 2d \int x_C \, dm + d^2 \int dm
$$

The first term of $ \int x_C^2 \, dm $ is the moment of inertia about the center of mass axis:

$$
I_C = \int x_C^2 \, dm
$$

The second term $ \int x_C \, dm $ corresponds to the first moment of mass about the center of mass, which is zero:

$$
\int x_C \, dm = 0
$$

The third term $ \int dm $ is the total mass of the body $ M $, so:

$$
I = I_C + M d^2
$$

which is Steiner's theorem.

---

## Applications of Steiner's theorem

Steiner's theorem is useful in calculating the moments of inertia of bodies in cases where the axis of rotation does not pass through the center of mass. For example:

- **A thin rod of length $ L $, rotating about its end:**
The moment of inertia about the center of the rod is:

$$
I_C = \frac{1}{12} M L^2
$$

If the axis is displaced by $ d = \frac{L}{2} $, then:

$$
I = \frac{1}{12} M L^2 + M \left(\frac{L}{2}\right)^2
$$

$$
I = \frac{1}{12} M L^2 + \frac{1}{4} M L^2 = \frac{1}{3} M L^2
$$

which corresponds to the well-known result for a rod rotating about one end.
