## Dynamics

> Dynamics is the branch of mechanics that studies the motion of bodies and the forces that cause or affect that motion.

Unlike kinematics, which describes motion without regard to its causes, dynamics examines the relationship between forces, mass, and acceleration.

### Force is a vector

Force is defined as a vector quantity for several key reasons, which stem from its nature and the way it affects the motion of bodies. Here are the main characteristics that define force as a vector:

1. **Magnitude** indicates how strong the interaction between bodies is. This magnitude is measured in newtons (N). The magnitude of a force tells us how much the force affects the motion of a body (for example, how fast it accelerates it).

2. **Direction** indicates in which direction the force acts. The direction of the force is key to determining how a body will move or deform. For example, if you apply a force to a body from the left, it will move to the right, and vice versa.

3. **The point of application** is the point at which a force acts on an object. The point of application of a force affects how the object will rotate or deform. For example, if you apply a force to the edge of a door, it will rotate, while if you apply a force to the center of the door, it will only slide.

4. **The vector of force** is the line on which the force vector lies. The vector of force is important for determining the moment of force and for adding forces.

#### Vector representation of force

A force is usually written as a vector $\vec{\mathbf{F}}$, which has components in different directions (for example, $F_x, F_y, F_z$ in Cartesian coordinates). This representation allows us to mathematically manipulate forces and perform calculations regarding their effects on bodies.

#### Sum of forces

When summing forces acting on a body, we try to find a resulting force that has the same effect as all the individual forces combined. There are numerical and graphical methods for determining this resulting force.

* **Numerical methods** use mathematical operations with force vectors to determine the resulting force.

1. Decomposition of forces into components

* We decompose each force into components in the Cartesian coordinate system ($F_x, F_y, F_z$).
* We add the components of the forces in the same direction:
* $\Sigma F_x = F_{1x} + F_{2x} + ... + F_{nx}$
* $\Sigma F_y = F_{1y} + F_{2y} + ... + F_{ny}$
* $\Sigma F_z = F_{1z} + F_{2z} + ... + F_{nz}$
* The resultant force ($\mathbf{F}$) has components ($\Sigma F_x$, $\Sigma F_y$, $\Sigma F_z$).
* The magnitude of the resultant force is calculated using the norm:
* $|\vec{\mathbf{F}}| = \sqrt{(\Sigma F_x)^2 + (\Sigma F_y)^2 + (\Sigma F_z)^2}$
* The direction of the resultant force is determined using trigonometric functions.

2. Vector sum
* Forces are added vectorially, which means that their effects are added according to the rules of vector sum.

* **Graphical methods** use geometric constructions to determine the resulting force.
1. Parallelogram rule
* Used for the sum of two forces.
* Forces are represented as vectors with a common origin.
* Vectors are added to a parallelogram.
* The resulting force is the diagonal of the parallelogram.

![parallelogram of forces](https://wiki.dtonline.org/images/thumb/e/e9/ParallelogramForces.png/300px-ParallelogramForces.png)

3. Triangle rule
* Used for the sum of two forces.
* Forces are represented as vectors such that the end of the first vector is the origin of the second vector.
* The resultant force is the vector that connects the origin of the first vector to the end of the second vector.

![triangle of forces](https://wiki.dtonline.org/images/thumb/c/cd/TriangleForces.png/300px-TriangleForces.png)

4. Polygon of forces
* Used for the sum of multiple forces.
* Forces are represented as vectors so that the end of one vector is the origin of the next vector.
* The resultant force is the vector that connects the origin of the first vector to the end of the last vector.

### Galileo law of inertia
A body at rest remains at rest and body in motion continues to move at constant velocity along a straight line unless acted upon by an external force.

### Newton's laws

1. **Newton's first law** - the law of inertia.

> *Corpus omne perseverare in statu suo quiescendi vel movendi uniformiter in directum, nisi quatenus illud a viribus impressis cogitur statum suum mutare.*

> If no external forces act on a body (matter), or the resultant of the forces is 0, then the body remains at rest or in uniform rectilinear motion.

The equivalent (understandable and literal) formulation is: A body remains at rest or in uniform rectilinear motion, unless it is forced to change this state by external forces. 

Newton's first law is valid only in **inertial frame of reference**.

>note: Frames of reference [wikipedia:](https://en.wikipedia.org/wiki/Frame_of_reference)
>
>In physics and astronomy, a frame of reference (or reference frame) is an abstract coordinate system, whose origin, orientation, and scale have been specified in physical space. It is based on a set of reference points, defined as geometric points whose position is identified both mathematically (with numerical coordinate values) and physically (signaled by conventional markers). An important special case is that of inertial reference frames, a stationary or uniformly moving frame.
>
>In classical physics and special relativity, an **inertial frame of reference** (also called an inertial space or a Galilean reference frame) is a frame of reference in which objects exhibit inertia: they remain at rest or in uniform motion relative to the frame until acted upon by external forces. In such a frame, the laws of nature can be observed without the need to correct for acceleration.

All frames of reference with zero acceleration are in a state of constant rectilinear motion (straight-line motion) with respect to one another. 

2. **Newton's Second Law** - the law of force

> *Mutationem motus proportionalem esse vi motrici impressae et fieri secundam lineam rectam qua vis illa imprimitur.*

> If a force acts on a body (mass), then the body moves with an acceleration that is directly proportional to the acting force and inversely proportional to the mass of the body.

or: A force action on a body gives it an acceleration which is in the direction of the force and has an magnitude givem by $m \vec{\mathbf{a}}$.

Newton's second law is also valid only in **inertial frame of reference**.

$$\vec{\mathbf{F}} = m \vec{\mathbf{a}} $$

$$\vec{\mathbf{F}} = m \ddot{\vec{\mathbf{x}}} $$

3. **Newton's Third Law** - the law of action and reaction

> *Actioni contrariam semper et aequalem esse reactionem; sive: corporum duorum actiones in se mutuo semper esse aequales et in partes contrarias dirigi.*

> Every action is always opposed by an equal but opposite reaction; otherwise: the mutual actions of two bodies are always equal and directed in opposite directions.

or: If one object exerts a force to another, the other exerts the same force in opposite direction on the one. This force is called a **contact force**.

## When is a force pair in equilibrium?

A force pair is in equilibrium when it meets the following three conditions:

1. **Equal magnitude:** Both forces must have the same magnitude (the magnitude of the force vector).

2. **Opposite direction:** The forces must act in opposite directions. This means that the force vectors must point in opposite directions.

3. **Common carrier:** The forces must act on the body in one straight line (their vectors must lie on one straight line).

### Consequences of equilibrium of a force pair

* **Zero resultant:** The vector sum of the two forces is zero. This means that the force pair has no translational effect on the body.
* **Zero moment:** The moment of the force pair is also zero. This means that the force pair has no rotational effect on the body.

### Examples

* **Book on a table:** The book is acted upon by a downward force of gravity and an equal upward force of the table. These two forces form a pair of forces in equilibrium.
* **A competitor holding a barbell:** The competitor exerts an upward force to balance the downward force of gravity on the barbell. These two forces form a pair of forces in equilibrium.

### Passive Forces in 1D

* **Friction ($\vec{\mathbf{F}}_t$)**: Friction is the force that occurs between two surfaces that are in contact and are moving against each other or trying to move. Friction always acts against the direction of motion or the direction of the effort to move.

* **Static friction ($\vec{\mathbf{F}}_{ts}$)**: Acts between bodies that are at rest and prevents them from moving.

$$ F_{ts} \leq \mu_s N$$

where $\mu_s$ is the coefficient of static friction and $N$ is the normal force.

* **Kinetic friction ($\vec{\mathbf{F}}_{td}$)**: Acts between bodies that are moving against each other.

$$ F_{tk} \leq \mu_k N$$

where $\mu_k$ is the coefficient of kinetic friction and $N$ is the normal force.

![friction](http://hyperphysics.phy-astr.gsu.edu/hbase/imgmec/fsta.gif)


* **Environmental drag (resistance) ($\vec{\mathbf{F}}_o$)**: Environmental resistance (e.g. air resistance or fluid resistance) is the force that opposes the motion of a body in a given medium. The magnitude of the environmental resistance depends on the speed of the body and on the properties of the medium.

* **Air drag (for low speeds)**:

$$F_o = -b v$$

where $b$ is the environmental resistance constant and $v$ is the body speed for small Reynolds numbers.

* **Air drag (for high speeds)**:

$$F_o = -1/2 \rho C_d A v^2$$

where $\rho$ is the density of the medium, $C_d$ is the resistance coefficient, $A$ is the area of ​​the body, and $v$ is the velocity of the body.

### Properties of Passive Forces

* **Acts against motion**: Passive forces always act against the direction of motion or the direction of the effort to move.
* **Depends on conditions**: The magnitude of passive forces often depends on conditions, such as the material of the surfaces (for friction), the speed of the body (for environmental resistance), or temperature.
* **Loss of energy**: Passive forces cause the loss of mechanical energy, which is converted into heat or other forms of energy.
