# Yield stress criterions and reduced stress under plane stress

Yield stress criterions are used to determine the limit state of a material, i.e. the moment when the material fails (plastic deformation, fracture). They are formulated based on various mechanical criteria and experimental observations.

> **Reduced stress** is an equivalent uniaxial stress that replaces a complex stress state and allows comparing multiaxial stress with the ultimate strength limit of the material.

$$\sigma_{\rm{red}} \leq \sigma_d $$

where
- $\sigma_{\rm{red}}$ is the reduced stress
- $\sigma_d$ is the allowable stress determined using a tensile test, e.g. yield strength

The most commonly used strength hypotheses are:

1. **Principal stress criterion (Rankin criterion)**
- The material fails if one of the principal stresses exceeds the tensile or compressive strength limit.
- Suitable for brittle materials.
- Criterion:

$$
\max(|\sigma_1|, |\sigma_2|, |\sigma_3|) \geq \sigma_{d}
$$

where $ \sigma_d $ is the allowable stress of the material.

2. **Maximum shear stress criterion (Tresco criterion)**
- The material fails if the maximum shear stress exceeds a critical value.
- Suitable for tough materials.
- Criterion:

$$
\tau_{\max} = \frac{\sigma_1 - \sigma_3}{2} \geq \tau_d
$$

where $ \tau_d $ is the ultimate shear stress.

3. **Maximum distortion energy criterion (Huber-Mises-Hencky hypothesis, HMH, vonMises)**
- The material fails if the critical value of the distortion energy is exceeded.
- The most commonly used hypothesis for ductile materials.
- In English written literature usualy as von Mises criterion
- Criterion:

$$
\sigma_{HMH} = \sqrt{\frac{(\sigma_1 - \sigma_2)^2 + (\sigma_2 - \sigma_3)^2 + (\sigma_3 - \sigma_1)^2}{2}} \geq \sigma_d
$$

## **Reduced stress under plane stress**
Under plane stress (e.g. a thin plate loaded in its plane) the third principal stress is zero, i.e. $ \sigma_3 = 0 $.

According to the Huba-Mises hypothesis, the reduced stress is:

$$
\sigma_{HMH} = \sqrt{\sigma_1^2 - \sigma_1\sigma_2 + \sigma_2^2}
$$

In case the shear stress is non-zero, we express the reduced stress as:

$$
\sigma_{HMH} = \sqrt{\sigma_x^2 + \sigma_y^2 - \sigma_x\sigma_y + 3\tau_{xy}^2}
$$

where:
- $ \sigma_x, \sigma_y $ are the normal stresses in the plane,
- $ \tau_{xy} $ is the shear stress.
