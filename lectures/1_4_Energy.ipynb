## Mechanical work

> When a body moves due to the action of a force, mechanical work is done.

**Notation:**
* Mechanical work is usually denoted by the letter **W** (from the English "work").

**Unit:**
* In the SI system (International System of Units), the unit of mechanical work is the **joule** (pronounced joule).
* The symbol for joule is **J**.

**Definition of joule:**
* 1 joule (1 J) is the work done by a force of 1 newton (1 N) when moving a body 1 meter (1 m) in the direction of the applied force.
* Mathematically: 1 J = 1 N ⋅ m

**Other units (less common):**
* **erg**: Used in the CGS (centimeter-gram-second) system. 1 erg = 10⁻⁷ J.
* **kilowatt-hour (kWh)**: Used to measure electrical energy consumption. 1 kWh = 3.6 × 10⁶ J.

1. **The force acts in the same direction as the body's motion**

$$W = F s $$

where $F$ is the magnitude of the force and $s$ is the path length

2. **The force acts in a different direction than the body's motion**

$$W = \vec{\mathbf{F}} \cdot \vec{\mathbf{s}} $$

where $\vec{\mathbf{F}}$ is the force vector and $\vec{\mathbf{s}}$ is the displacement vector. In other words, the work is done by the component of the force parallel to the body's trajectory. In the case where the force is constant and the trajectory is a straight line, we can express it as.

$$W = F s \cos \alpha$$

where $\alpha$ is the angle between the force and the trajectory of the motion.

3. **The force is changing or the path is curved**

$$W = \int\limits_0^s \vec{\mathbf{F}} \cdot \mathrm{d}\vec{\mathbf{s}} $$

## Energy

### Kinetic Energy
Kinetic energy is the energy associated with the motion of an object. A moving object has kinetic energy. The kinetic energy of a particle of mass m moving at a velocity v is given by:
$$E_k = \frac{1}{2} m \vec{\mathbf{v}} \cdot \vec{\mathbf{v}}= \frac{1}{2} m v^2$$

where $v=|∣\vec{\mathbf{v}}∣|$ is the magnitude of the velocity. For a system of particles, the total kinetic energy is the sum of the kinetic energies of all individual particles:
$$E_k = \sum_i \frac{1}{2} m_i \dot{\vec{\mathbf{r}}}_i \cdot \dot{\vec{\mathbf{r}}}_i = \sum_i \frac{1}{2} m_i v_i^2$$
where $m_i$ is the mass of the $i$-th particle and $\dot{\vec{\mathbf{r}}_i}$ is its velocity

### Potential Energy

Potential energy is the energy associated with the configuration of a system of objects. It represents the ability to do work due to the position of the object in a force field. Potential energy is often associated with conservative forces. A force $F$ is conservative if the work done by that force on a particle moving between two points does not depend on the path.

The potential energy $U$ associated with a conservative force $F$ is defined as:
$$\mathbf{\vec{F}} = -\nabla U$$
where $\nabla$ is the gradient operator. This means that the force is the negative gradient of the potential energy. The change in potential energy between two points is equal to the negative work done by the conservative force in moving between those points.

### Conservative force [wikipedia](https://en.wikipedia.org/wiki/Conservative_force#Non-conservative_force)
In physics, a conservative force is a force with the property that the total work done by the force in moving a particle between two points is independent of the path taken. Equivalently, if a particle travels in a closed loop, the total work done (the sum of the force acting along the path multiplied by the displacement) by a conservative force is zero.

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/1/1e/Conservative_Force_Gravity_Example.svg/179px-Conservative_Force_Gravity_Example.svg.png)

Common examples of potential energy include:
* **Gravitational potential energy:** For an object of mass m near the surface of the Earth, the gravitational potential energy is:
$$E_p = mgh$$
where $g$ is the acceleration due to gravity and $h$ is the height above the reference level.

* **Elastic Potential Energy:** For a spring with stiffness constant $k$ stretched or compressed by a distance $x$ from its equilibrium position, the elastic potential energy is:
$$U_e = \frac{1}{2} k x^2$$

### Work and Energy

The work $W$ done by the resultant force on a particle as it moves from point A to point B is equal to the change in its kinetic energy:
$$ W = \Delta T = \Delta E_k = E_{kB} - E_{kA} = \frac{1}{2} m v_B^2  - \frac{1}{2} m v_A^2$$
If only conservative forces do the work, then we can also combine the work with the change in potential energy:
$$ W = -\Delta U = -(U_B - U_A) = U_A - U_B$$
Combining these relations, we obtain the law of conservation of mechanical energy:
$$\Delta T + \Delta U = 0 \quad \text{or} \quad T_A + U_A =T_B + U_B$$
This means that the total the mechanical energy ($E=T+U=\frac{1}{2} m v^2 + mgh$) of a system remains constant if only conservative forces do work.

## Power
Power ($P$) is defined as the rate at which work is done or energy is transferred:

$$P = \frac{dW}{dt} = \vec{\mathbf{F}} \cdot \vec{\mathbf{v}}$$

Power is measured in watts (W), where 1 W = 1 J/s.
