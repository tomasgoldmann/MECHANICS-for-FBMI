# Buckling

> **Buckling failure** is the sudden deformation of a structural member under compression that occurs when the compressive load on the member reaches a critical value.

Buckling often occurs suddenly and can lead to large displacements. Although material failure (cracking or failure) does not always occur, buckling is considered a **failure** because the buckling structure can no longer perform its original function.

![A structural member buckles under compression](https://efficientengineer.com/wp-content/uploads/buckling-failure.jpg)

### Euler's equation for buckling

The problem of the stability of an axially loaded member was solved as early as 1744 by Leonard Euler
(1707 – 1783). Engineers initially ignored his findings—until several truss railway bridges collapsed in the early stages of railway development. (The worst accident occurred in the mid-19th century in Switzerland, Euler’s native country.) Since then, strict standards have been in place for checking the buckling strength of axially compressed members.

Euler’s equation is used to calculate the axial load $F_\mathrm{krit}$ at which a member or beam loses stability. We call this force the critical force at the buckling limit. At the critical load, any small deviation, whether it is a lateral force or a small imperfection in the member’s geometry, will cause a loss of stability.

$$F_\mathrm{krit} = \frac{\pi^2 EJ_{\mathrm{min}}}{l_\mathrm{red}^2}$$

The critical load depends on only three parameters:
* **Young's modulus** $E$ of the material
* **square moment of section** $J_\mathrm{min}$ -- in the case of different principal square moments, the smaller one is taken
* **reduced length** $l_\mathrm{red}$ of the member.

### Reduced Length and Boundary Conditions

The way in which the ends of a member or beam are restrained will affect the critical buckling load. A member that is rigidly supported at one end and free at the other will obviously be able to support a much smaller load before buckling compared to one that is pinned at both ends. The buckling shapes are also quite different.

![Critical Buckling Load of a Column Pinned at Both Ends and a Column Fixed at One End](https://efficientengineer.com/wp-content/uploads/buckling_end_conditions.jpg)

Euler's formula takes into account different end conditions using the **reduced length** parameter $l_\mathrm{red}$.

A member that is pinned at both ends has an effective length that is equal to the length of the column. Other end conditions result in an effective length that is equal to the length multiplied by a specific factor. These factors are listed in the table below for several different end conditions.

| End conditions | Effective length |
| :-------------- | :---------------- |
| Pinned at both ends | $l = l_\mathrm{red}$ |
| Pinned at one end, free at the other | $l_\mathrm{red} = 2 l$ |
| Pinned at both ends | $l_\mathrm{red} = 0.5 l$ |
| Pinned at one end, hinged at the other | $l_\mathrm{red} = 0.7 l$ |

### Slenderness of a member

Even without knowing Euler's formula, it is intuitively obvious that slender columns are much more susceptible to buckling than thick ones. Therefore, members of a truss structure that are loaded in compression are sometimes designed to be stronger than those loaded in tension, and stiffeners are used to prevent buckling of long compression members.

Euler's formula confirms this intuition - the length term in the equation is squared, and as a result, doubling the length of the column means that it can only support a quarter of the load before buckling.

To better understand the effect of slenderness, it is useful to introduce a dimensionless parameter called **slenderness $\lambda$**, which defines how slender the column is (i.e., how thin it is compared to its length).

$$\lambda = \frac{l_{red}}{\sqrt{\frac{J_{min}}{A}}} $$

A real member has a finite value of the yield strength. This raises the question of whether the elastic limit state or the buckling stability state occurs earlier for a given member. Dividing the Euler critical force by the cross-sectional area of ​​the member gives:

$$\sigma_\mathrm{krit} = \frac{F_\mathrm{krit}}{A} = \pi^2 \frac{E J_{min}}{A l_\mathrm{red}^2} = \pi^2 \frac{E}{\lambda^2}$$

As mentioned earlier, it is intuitive that the slenderness of a column will significantly affect how much load it can support before buckling, so it is interesting to see how the Euler critical buckling stress varies with slenderness ratio - this is shown in the graph below.

![Critical buckling stress vs. slenderness ratio](https://efficientengineer.com/wp-content/uploads/slenderness-ratio.jpg)

Very slender columns have a high slenderness ratio and a very low critical buckling stress. For thick columns with low slenderness ratios, the critical buckling stress will be very high.

The dashed horizontal line in the graph above shows the yield strength of the column material in compression $\sigma_K$. For very low slenderness ratios, the ultimate strength of the material will be exceeded before the buckling limit is reached. This means that we can define two distinct regions, one where the columns fail by crushing because the stresses in the column exceed the yield strength of the material (i.e., for thick columns), and one where they fail by buckling (i.e., for slender columns).

This curve only represents the theoretical behavior of the columns. If we plot the experimentally determined buckling stresses for real columns, we will see that it does not exactly match the theoretical behavior - this is shown in the graph below. In particular, the transition between plastic failure (crushing) and elastic failure (buckling) is much smoother. This is because for columns in this transition range, buckling is actually a complex combination of these two failure modes. Buckling in this transition region is called **inelastic buckling** and should be modeled using methods such as **Engesser theory** or **Shanley theory** rather than Euler's formula.

![Difference between experimental and theoretical results in the inelastic buckling region](https://efficientengineer.com/wp-content/uploads/inelastic-buckling.jpg)

Many industry design standards include curves similar to those shown above that can be used for the design of members loaded in compression.
