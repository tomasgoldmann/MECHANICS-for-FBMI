## Uniform Circular Motion

> Uniform circular motion is a specific type of motion in which an object moves in a circle at a constant speed.

For example, any point on a propeller rotating at a constant speed is in uniform circular motion. Other examples are the second, minute, and hour hands of a watch. It is noteworthy that the points on these rotating objects actually accelerate, even though the speed of rotation is constant. To understand this, we need to analyze the motion using vectors.

### Parameters of Uniform Circular Motion (1DOF - Scalar Description)

* **Angular Position ($\varphi$):**
* Indicates the position of an object on a circle relative to a reference point (usually the origin).
* Measured in radians (rad).
* The change in angular position with time describes the motion of an object in a circle.
* **Angular Velocity ($\omega$):**
* Indicates the change in angular position per unit of time.

$$\omega = \frac{d\varphi}{dt}$$

* Measured in radians per second (rad/s).
* When moving uniformly around a circle, the angular velocity is constant.
* **Frequency ($f$):**
* Indicates the number of revolutions per unit time.

$$f = \frac{\omega}{2\pi}$$

* Measured in hertz (Hz).
* Frequency is the reciprocal of the period.
* **Period ($T$):**
* Indicates the time required to complete one revolution.

$$T = \frac{1}{f} = \frac{2\pi}{\omega}$$

* Measured in seconds (s).

* **Circular velocity ($v$):**

$$v = r\omega$$

, where $r$ is the radius of the circle.

* The circumferential velocity is tangent to the trajectory.

The motion of a body can be described in terms of angular position as a function of time:

* $\varphi(t) = \varphi_0 + \omega t$

Where:

* $\varphi(t)$ is the angular position at time $t$
* $\varphi_0$ is the initial angular position (phase of motion)
* $\omega$ is the angular velocity

### Centripetal acceleration

In one-dimensional kinematics, objects with constant velocity have zero acceleration. However, in two- and three-dimensional kinematics, even if the velocity is constant, a particle can have an acceleration if it moves along a curved path, such as a circle. In this case, the velocity vector changes, or

$$ d\vec{v}\text{/}dt\ne 0. $$

As a particle moves counterclockwise at time $\Delta t$ along a circular path, its position vector shifts from $ \vec{r}(t)$ to $\vec{r}(t+\Delta t)$. The velocity vector has a constant magnitude and is tangent to the path, while when it changes from $\vec{v}(t)$ to $\vec{v}(t+\Delta t)$, only its direction changes. Since the velocity vector $\vec{v}(t)$ is perpendicular to the position vector $\vec{r}(t)$, the triangles formed by the position vectors and $\Delta\vec{r}$ and the velocity vectors $\Delta \vec{v}$ are similar. Moreover, since $ |\vec {r}(t)|=|\vec{r}(t+\Delta t)|$ and $|\vec{v}(t)|=|\vec{v}(t+\Delta t)|$, both triangles are isosceles. From these facts we can make the statement

$$\frac{\Delta v}{v}=\frac{\Delta r}{r} $$

or

$$ \Delta v=\frac{v}{r}\Delta r$$

The magnitude of the acceleration can be found from:
$$a=\underset{\Delta t\to 0}{\text{lim}}\left(\frac{\Delta v}{\Delta t}\right)=\frac{v}{r}\left(\underset{\Delta t\to 0}{\text{lim}}\frac{\text{Δ}r}{\text{Δ}t}\right)=\frac{{v}^{2}}{r}.$$

The direction of the acceleration can also be found by observing that as $ \text{Δ}t $ and therefore $ \text{Δ}\theta $ approaches zero, the vector $ \text{Δ}\overset{\to }{v} $ approaches the direction perpendicular to $ \overset{\to }{v}. $ In the limit $ \text{Δ}t\to 0,$ $ \text{Δ}\overset{\to }{v} $ is perpendicular to $ \overset{\to }{v}. $ Since $ \overset{\to }{v} $ is tangent to the circle, the acceleration $ \frac{d\overset{\to }{v}}{dt} $ is directed toward the center of the circle. In summary, a particle moving in a circle at a constant velocity has an acceleration of the magnitude

$${a}_{\text{C}}=\frac{{v}^{2}}{r}.$$

The direction of the acceleration vector is toward the center of the circle. This is the radial acceleration and is called the centripetal acceleration, which is why we give it the subscript c.

> **note**
>
> The word centripetal comes from the Latin words *centrum* (meaning "center") and *petere* (meaning "to seek"), and thus has the meaning of "seeking the center".

## Non-uniform circular motion

Circular motion is not necessarily a constant velocity. A particle can move in a circle and accelerate or decelerate, which shows acceleration in the direction of motion.

In uniform circular motion, a particle moving in a circle has a constant velocity and the circle has a fixed radius. If the velocity of the particle is also changing, then we introduce an additional acceleration in the direction tangent to the circle. Such accelerations occur at a point on a flywheel that is changing its speed, or in any accelerating rotor. If the velocity of a particle is changing, then it has a **tangent acceleration**, which is the time rate of change of the magnitude of the velocity:

$$ {a}_{\text{T}}=\frac{d|\overset{\to }{v}|}{dt}. $$

The direction of tangential acceleration is tangent to the circle, while the direction of centripetal acceleration is radially inward toward the center of the circle. A particle in circular motion with tangential acceleration therefore has a **total acceleration**, which is the vector sum of the centripetal and tangential accelerations:

$$ \overset{\to }{a}={\overset{\to }{a}}_{\text{C}}+{\overset{\to }{a}}_{\text{T}}. $$

The acceleration vectors are shown in (Figure). Note that the two acceleration vectors $ {\overset{\to }{a}}_{\text{C}} $ and $ {\overset{\to }{a}}_{\text{T}} $ are perpendicular to each other, with $ {\overset{\to }{a}}_{\text{C}} $ in the radial direction and $ {\overset{\to }{a}}_{\text{T}} $ in the tangential direction. The total acceleration $ \overset{\to }{a} $ is directed at an angle between $ {\overset{\to }{a}}_{\text{C}} $ and $ {\overset{\to }{a}}_{\text{T}}.$
