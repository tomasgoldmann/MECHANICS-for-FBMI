# Periodic motion

## Periodic motion and its characteristics
![](https://labs.phys.utk.edu/mbreinig/phys221core/modules/m11/images/shmintro.gif)![](https://labs.phys.utk.edu/mbreinig/phys221core/modules/m11/images/shm-animation.gif)
> Periodic motion is a type of motion in which a body or system returns to the same state after a certain period of time.

The time required for one complete repetition of the motion is called **period** and is usually denoted by the letter $T$. The symbol $P$ is not used because it could be confused with momentum. One complete repetition of the motion is called **cycle**.
The unit of period in the **SI system** is the second **(s)**.

Mathematically, periodic motion can be described by the function:

$$x(t+T)=x(t)$$

where where $x(t)$ represents the position describing the motion at time $t$

### Frequency

> **Frequency** is defined as the number of cycles per unit time.

It is usually denoted by the Latin letter $f$ or the Greek letter $\nu$ (*ni*). Period and frequency are **reciprocal quantities**:

$$
f = \frac{1}{T}
$$

For example, if a newborn's heart beats at **120 beats per minute**, then its period (the time between two beats) is **0.5 seconds**. In general, **a higher frequency corresponds to a shorter period** and vice versa.

In the **SI system**, the unit of frequency is **hertz (Hz)**, named after the German physicist Heinrich Hertz. **1 Hz** means that a given event repeats **once per second**.

For rotating devices, the traditional unit **revolutions per minute (RPM)** is often used:

$$
60 \text{ RPM} = 1 \text{ Hz}
$$

## Harmonic periodic motion

> Harmonic periodic motion is a special type of periodic motion in which the position of a body over time can be described by a sine or cosine function.

$$x(t) = A \cos \left(\omega t + \varphi_0 \right)$$

where
* $x(t)$ is the instantaneous position of the oscillating body
* $t$ is time
* $A$ is **amplitude** – the maximum deviation from the equilibrium position,
* $\omega,\varphi_0$ are constants describing the motion

### Angular frequency

Periodic motion is often better described by **angular frequency**, which is denoted by the Greek letter $\omega$ (*omega*). Angular frequency represents **angular displacement per unit time** (for example, during rotation) or the rate of change of phase of a sinusoidal signal.

$$
y(t) = \sin(\varphi(t)) = \cos(\omega t) = \cos(2\pi f t)
$$

Formula for angular frequency:

$$
\omega = 2\pi f \quad
$$

Angular frequency is usually expressed in units of **radians per second** (**rad/s**), with one full circle containing **2 $\pi$ radians**.

### Harmonic phase

> **Harmonic phase** determines the instantaneous state of the oscillator at a given time \( t \). It is expressed as the argument of the trigonometric function in the equation of harmonic motion:

$$
\varphi(t) = \omega t + \varphi_0
$$

where:
- $\varphi(t)$ is the **instantaneous phase** of the oscillator,
- $\omega$ is the **angular frequency**,
- $t$ is the **time**,
- $\varphi_0$ is the **initial phase**, which determines the state of the oscillator at time $t = 0$.

The phase of the harmonic motion determines **in which part of the cycle the oscillator is**.
- If $\varphi_0 = 0$, the oscillator starts at **maximum deflection** (moving according to the cosine function).
- If $\varphi_0 = \frac{\pi}{2}$, the oscillator starts at **equilibrium position** and moves upwards.

Harmonic motion can be expressed as:

$$
x(t) = A \cos(\omega t + \varphi_0)
$$

The phase $\varphi(t) = \omega t + \varphi_0$ increases **linearly in time**, which corresponds to the periodicity of the motion. The phase is measured in **radians** or **degrees**:
- If the phase reaches $2\pi$, the oscillator has completed one complete cycle.

## Relationship between harmonic motion and circular motion

Harmonic motion can be understood as the **projection of circular motion** into one of its coordinate directions. Let us imagine a mass point that moves **uniformly around a circle** with an angular velocity $\omega$. Its position can be described using parametric equations:

$$
x(t) = R \cos(\omega t + \varphi_0)
$$

$$
y(t) = R \sin(\omega t + \varphi_0)
$$

If we only observe the coordinate $x(t)$, we see that it behaves like a harmonic oscillator with amplitude $A = R$.

## Harmonic Oscillator Velocity
Velocity is the first time derivative of position:

$$
v(t) = \dot{x} = \frac{dx(t)}{dt} = - A \omega \sin(\omega t + \varphi_0)
$$

- **The amplitude of the velocity** is $v_{\max} = A \omega$.
- The velocity is **phase shifted** by $\frac{\pi}{2}$ from the position – when the deflection is maximum, the velocity is zero.
- The sign of the velocity depends on the direction of motion:
- When moving **towards the equilibrium position**, the velocity is positive or negative depending on the direction.
- When moving **away from the equilibrium position**, it has the opposite sign.

## Acceleration of a harmonic oscillator
Acceleration is the first derivative of velocity, i.e. the second derivative of position:

$$
a(t) = \ddot{x} = \frac{dv(t)}{dt} = - A \omega^2 \cos(\omega t + \varphi_0)
$$

- **The amplitude of acceleration** is $a_{\max} = A \omega^2$.
- The acceleration is **phase shifted by $\pi$** compared to the position – it always points towards the equilibrium position.
- The sign of acceleration:
- It is **positive** when the deflection is negative $x < 0$ – it points back towards the equilibrium position.
- It is **negative** when the deflection is positive $x > 0$ – it points back towards the equilibrium position again.

## Concepts of kinematics of periodic motion

| **Quantity** | **Designation** | **Unit (SI)** | **Description** |
|-------------------------|----------------|-----------|----------|
| **Period** | $T$ | second (s) | Time required for one cycle of motion. |
| **Frequency** | $f$ or $\nu$ | hertz (Hz) | Number of cycles per second. Relates to period as $f = \frac{1}{T}$. |
| **Angular frequency** | $\omega$ | rad/s | Indicates the rate of change of phase of the oscillator: $\omega = 2\pi f$. |
| **Amplitude** | $A$ | meter (m) | Maximum deviation from equilibrium. |
| **Phase** | $\varphi(t)$ | radian (rad) | Instantaneous state of the oscillator: $\varphi(t) = \omega t + \varphi_0$. |
| **Initial phase** | $\varphi_0$ | radian (rad) | Phase shift at $t = 0$. |
