# Static equations of equilibrium in a plane

Mechanical equilibrium occurs when a body or system of bodies is at rest. Static equations of equilibrium are the basic tools for analyzing the equilibrium of rigid bodies in a plane. These equations express the conditions that must be met for the body to remain at rest.

## Vector notation

The vector notation of the static equations of equilibrium expresses that the resultant of all forces and the resultant moment of the forces acting on the body must be zero.

$$\sum \vec{F} = \vec{0}$$

$$\sum \vec{M}_O = \vec{0}$$

Where:

* $\sum \vec{F}$ is the vector sum of all forces acting on the body.
* $\sum \vec{M}_O$ is the vector sum of the moments of all forces with respect to point O.

## Component notation

In the plane, we can decompose vector equations into components in the Cartesian coordinate system (x and y axes).

$$\sum F_x = 0$$

$$\sum F_y = 0$$

$$\sum M_O = 0$$

Where:

* $\sum F_x$ is the algebraic sum of the components of the forces in the x-axis direction.
* $\sum F_y$ is the algebraic sum of the components of the forces in the y-axis direction.
* $\sum M_O$ is the algebraic sum of the moments of the forces with respect to point O.

## Meaning of the equations

* **Balance of forces:** The first two equations ($\sum F_x = 0$ and $\sum F_y = 0$) express the condition that the body does not move in translation.
* **Balance of moments:** The third equation ($\sum M_O = 0$) expresses the condition that the body does not rotate.

## Uses

Static equilibrium equations are used to solve many problems in statics, such as:

* Determining the reactions of connections.

* Analyzing beams, frames, and other structures.
* Calculating the forces acting on bodies in equilibrium.

## Sign convention

When solving equations, it is important to follow the sign convention:

* Components of forces in the direction of the positive x-axis are positive, in the opposite direction are negative.
* Components of forces in the direction of the positive y-axis are positive, in the opposite direction are negative.
* Moments of forces that cause counterclockwise rotation are positive, in the clockwise direction are negative.

## Equilibrium states

Equilibrium states are divided according to how the body behaves when it is slightly displaced from its equilibrium position:

1. **Stable equilibrium:**
* When it is slightly displaced from its equilibrium position, the body returns to its original position.
* Example: A ball in a hole.
2. **Unstable equilibrium:**
* When it is slightly displaced from its equilibrium position, the body moves away from its original position.
* Example: A ball on top of a hill.
3. **Indifferent equilibrium:**
* When it is slightly displaced from its equilibrium position, the body remains in its new position.
* Example: A ball on a horizontal plane.

### Factors affecting equilibrium states

* **Position of the center of gravity:** The position of the center of gravity of a body plays a key role in determining stability.
* **Support:** The way a body is supported affects its stability.
* **External forces:** External forces acting on a body can affect its state of equilibrium.
